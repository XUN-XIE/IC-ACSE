## Homework - Fibonacci matrix

The Fibonacci matrix is defined as

$$Q = \begin{pmatrix}
1 & 1\\
1 & 0
\end{pmatrix}$$

Compute by hand the eigenvalues and eigenvectors of the $Q$ matrix.

Check numerically that you can use these to form a diagonalization of $Q$.

We will make use of this matrix and its diagonalisation in lecture 4.

### Solution - Fibonacci matrix

The eigenvalues of the matrix $Q$ are given by 

$$0 = |A-\lambda I| = 
\begin{vmatrix}
1-\lambda & 1\\
1 & -\lambda
\end{vmatrix} = 
\lambda^2 - \lambda -1$$

which has the roots

$$\lambda_1 = \frac{1+\sqrt{5}}{2},\qquad \lambda_2 =  \frac{1-\sqrt{5}}{2}$$

to find the eigenvectors let's solve

$$
\begin{align*}
\boldsymbol{0} & = (A-\lambda I)\boldsymbol{v}\\[10pt]
 & =
\begin{pmatrix}
1-\lambda & 1\\
1 & -\lambda
\end{pmatrix}\boldsymbol{v}
\end{align*}
$$

We could perform row operations to solve this system, but also we can notice that the solution is

$$\boldsymbol{v} = \begin{pmatrix}
\lambda\\
1
\end{pmatrix}$$

Now since our eigenvalues are distinct,

$$\boldsymbol{v}_1 = \begin{pmatrix}
\lambda_1\\
1
\end{pmatrix},\qquad
\boldsymbol{v}_2 = \begin{pmatrix}
\lambda_2\\
1
\end{pmatrix}
$$

are two linearly independent eigenvectors. We won't bother normalising them as it doesn't matter for our application.

Hence,

$$P = \begin{pmatrix}
\lambda_1 & \lambda_2\\
1 & 1
\end{pmatrix},\qquad
\Gamma =
\begin{pmatrix}
\lambda_1 & 0\\
0 & -\lambda_2
\end{pmatrix}$$

Let's check this

````python
Q = np.array([[1., 1.], [1.,0.]])
lam1 = (1+np.sqrt(5))/2
lam2 = (1-np.sqrt(5))/2
v1 = np.array([lam1, 1.])
v2 = np.array([lam2, 1.])

# force v1 and v2 to be column vecs
v1.shape = (2,1)
v2.shape = (2,1)

# and stack them into the matrix P
P = np.hstack((v1,v2))
# and form the diagonal matrix with e'vals on the main diagonal
Lambda = np.diag(np.array([lam1,lam2]))

# we can check if our result is (approximately - read the docs) correct with
print('Q = P \Lambda P^{-1}:  ', np.allclose(Q , P@Lambda@(sl.inv(P))))
````

## Homework - eigenvalues of the MSD system

Find the eigenvalues of the matrix that corresponds to the MSD matrix from the lecture:

$$
\begin{pmatrix}
0 & 1\\
-\omega_0^2 & -2\zeta\omega_0
\end{pmatrix}.$$

### Solution - eigenvalues of the MSD system


We can find the eigenvalues for our MSD system by finding the roots of the [characteristic equation](https://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors#Eigenvalues_and_the_characteristic_polynomial), i.e. solving the algebraic equation

$$\det(A - \lambda I) = 0,$$

for $\lambda$.

<br>

Let's do this for the $A$ for the MSD problem:

$$\det(A - \lambda I) = 
\begin{vmatrix}
-\lambda & 1\\
-\omega_0^2 & -2\zeta\omega_0 - \lambda
\end{vmatrix}
= \lambda^2 + 2\zeta\omega_0 \lambda + \omega_0^2. $$


The eigenvalues are given by the roots of this quadratic. Recall that a formula for the two roots of a quadratic in the form $ax^2+bx+c=0$ is

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

Applying this in our case yields the two eigenvalues

$$\lambda_1 = \omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right),\;\;\;\;
\lambda_2 = \omega_0\left(-\zeta - \sqrt{\zeta^2-1}\right).$$


Note that we have three cases:


1. $\zeta=0$: purely imaginary eigenvalues


2. $0<\zeta<1$: complex eigenvalues with $Re(\lambda)<0$


3. $\zeta\ge 1$: real (negative) eigenvalues



$\zeta$ here is encoding something about the physics of our problem (refer to the linked videos above and <https://en.wikipedia.org/wiki/Harmonic_oscillator#Damped_harmonic_oscillator>).

In the three cases above the system is termed: *undamped*, *under-damped* and *over-damped* (with $\zeta=1$ *critically-damped*), respectively.

## Homework - eigenvectors of the MSD system [**]

Work out expressions for the eigenvectors of the system.

### Solution - eigenvectors of the MSD system

To find the corresponding eigenvectors we need to solve the system

$$
\boldsymbol{0} = (A - \lambda I)\boldsymbol{v} = 
\begin{pmatrix}
-\lambda & 1\\
-\omega_0^2 & -2\zeta\omega_0 - \lambda
\end{pmatrix}\boldsymbol{v}
$$

where 

$$\lambda_1 = \omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right),\;\;\;\;
\lambda_2 = \omega_0\left(-\zeta - \sqrt{\zeta^2-1}\right).$$

Consider $\lambda_1$

$$
\begin{align*}
\boldsymbol{0} & =
\begin{pmatrix}
-\lambda & 1\\
-\omega_0^2 & -2\zeta\omega_0 - \lambda
\end{pmatrix}\boldsymbol{v}\\[10pt]
& =
\begin{pmatrix}
-\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1\\
-\omega_0^2 & -2\zeta\omega_0 -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right)
\end{pmatrix}\boldsymbol{v}\\[10pt]
& =
\begin{pmatrix}
-\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1\\
-\omega_0^2 & -\omega_0 \left(\zeta + \sqrt{\zeta^2-1}\right)
\end{pmatrix}\boldsymbol{v}
\end{align*}
$$

Let's form the augmented matrix and use row operations/Gassian elimination to solve this problem


\begin{align*}
[A \, | \, \boldsymbol{b}] = 
&
\left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   -\omega_0^2 & -\omega_0 \left(\zeta + \sqrt{\zeta^2-1}\right) & 0 
  \end{array}
\right]\\[10pt]
\xrightarrow{(2)\leftarrow (2)\times \left(-\zeta + \sqrt{\zeta^2-1}\right)/\omega_0}
&
\left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & -\left(\zeta + \sqrt{\zeta^2-1}\right)\left(-\zeta + \sqrt{\zeta^2-1}\right) & 0
  \end{array}
\right]\\[10pt]
= & \left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0
  \end{array}
\right]\\[10pt]
& \qquad\text{[as $-\left(\zeta + \sqrt{\zeta^2-1}\right)\left(-\zeta + \sqrt{\zeta^2-1}\right)= \zeta^2 - (\zeta^2-1) = 1$]}\\[10pt]
\xrightarrow{(2)\leftarrow (2)-1}
&
\left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   0 & 0 & 0
  \end{array}
\right]
\end{align*}

This shows that this system only has one linearly independent row, hence the original matrix ($A-\lambda I$) is singlular. Of course this is exactly as expected based on how we define our eigenvalues in the first place, and so for this $2\times 2$ problem we could have just considered one of the rows immediately without going through these row operations, for larger problems however we would need to do this in general.

What this exercise allows us to do is find $\boldsymbol{v}$ - the same $\boldsymbol{v}$ solves the original problem
$A\boldsymbol{v}=\lambda \boldsymbol{v}$ as solves the transformed problem encoded in the augmented system above:

$$
\left[
  \begin{array}{cc}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1   \\
   0 & 0
  \end{array}
\right]
\boldsymbol{v} = \boldsymbol{0}$$

This tells us that 

$$-\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) v_1 + v_2 = 0
\iff v2 = \omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) v_1 
$$

we therefore have a family of possible solutions which can be written as

$$\boldsymbol{v} = \alpha
\begin{pmatrix}
1\\
\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right)
\end{pmatrix}$$

for any scalar $\alpha$.  We can choose this to make this a unit vector:

$$\alpha = \frac{1}{\sqrt{1+\omega_0^2\left(-\zeta + \sqrt{\zeta^2-1}\right)\overline{\left[\left(-\zeta + \sqrt{\zeta^2-1}\right) \right]}}}
$$
where the overline indicates complex conjugate.

I ran out of steam and didn't compute the eigenvector for the second e'val.

Let's check if what we've computed for the first agrees with SciPy:

````python
# some parameters for problem
om0 = 1
zeta = 1.1 + 0j

A = np.matrix([[0, 1], [-om0**2, -2*zeta*om0]])

lam, vecs = sl.eig(A)

# check the first eigenvalues
print(om0*(-zeta + np.sqrt(zeta**2 - 1)))
print(lam[0])


# my normalised eigenvector
vec1 = np.array([1, om0*(-zeta + np.sqrt(zeta**2 - 1))])
print(vec1 / sl.norm(vec1))

# what scipy thinks
print('vec1:',vecs[:,0])
print('vec2:',vecs[:,1])
````

### Solution - stability of forward Euler applied to the MSD problem

Here we implement a solver and a function for MSD. First we check it gives the same plot as in the lecture (note the "eyeball norm" is a weak norm to use for code verification and you should do this more rigorously by directly comparing code output).

Then we see what happens in the cases of the physical/numerical parameters given in the lecture.

````python
def forward_euler(f, u0, t0, t_max, dt):
    """ Forward Euler time-stepper.
    f = f(t,y) is the RHS function.
    u0 is the initial condition.
    t0 is the initial time; t_max is the end time.
    dt is the time step size.
    """ 
    u = np.array(u0)
    t = np.array(t0)
    u_all = [u0]
    t_all = [t0]
    while t < t_max:
        u = u + dt*f(t, u)  # euler guess
        u_all.append(u)
        t = t + dt
        t_all.append(t)
    return np.array(u_all), np.array(t_all)

def f_msd(t, u):
    """ RHS function for mass-spring-damper problem.
    
    u for MSD is the vector [x, v]
    """
    f = np.zeros_like(u)
    f[0] = u[1]
    f[1] = -2.*zeta*om0*u[1] - om0**2 * u[0]
# NB. could write this as A*u with
# A = np.matrix([[0, 1], [-om0**2, -2*zeta*om0]])
    return f

# Let's first check this against the implementation from class - eyeball norm a plot

# numerical parameters for problem
om0 = 1
zeta = 0.1
# initial data (we need two for this coupled problem)
x0 = 1
v0 = 0
# simulation period and time step size
t0 = 0.
tf = 20.
dt = 0.04

# initial condition
u0 = np.array([x0, v0])
uall, tall = forward_euler(f_msd, u0, t0, tf, dt)


# pull out the individual solution vectors (don't need to do this)
x = uall[:, 0]
v = uall[:, 1]

fig = plt.figure(figsize=(8, 6))
ax1 = plt.subplot(111)
ax1.plot(tall, x, 'b', label='x')
ax1.plot(tall, v, 'r', label='v')
ax1.set_xlabel(r'$t$', fontsize=14)
ax1.set_ylabel(r'$x,\; v$', fontsize=14)
ax1.set_title('Numerical vs exact solution', fontsize=16)
ax1.grid(True)


def msd_exact_sol(t, zeta, om0):
    """Exact solution to the general MSD problem 
    Return: x and v component of solution
    """
    om = om0 * np.sqrt(1-zeta**2)
    phi = np.arctan2( (-v0 - om0*zeta*x0) , (om*x0) )
    CC = x0/(2*np.cos(phi))
    a = -om0*zeta
    b = om0 * np.sqrt(1-zeta**2)
    x_exact = 2*CC*np.exp(a*t)*np.cos(b*t+phi)
    v_exact = 2*CC*np.exp(a*t)*(-b*np.sin(b*t+phi) + a*np.cos(b*t+phi))
    return x_exact, v_exact


tfine = np.arange(0, tf, dt/100)
x_exact, v_exact = msd_exact_sol(tfine, zeta, om0)
ax1.plot(tfine, x_exact, 'k', label='exact x solution')
ax1.plot(tfine, v_exact, 'g', label='exact v solution')
ax1.legend(loc='best', fontsize=14);
````

- So for the first combination of parameters, the exact solution is damped a certain amount physically, the numerical solution is not damped as much, but still appears to result in a stable solution (appears as we haven't proved this, nor run for very long - you could test running longer if you want).


- For the second set, we increased the time step size, but kept the physics unchanged, this led to instability in the numerics while the true solution is stable.


- For the third, we kept this unstable time step size, but changed the physics of the problem, note that we now recover a stable calculation. This emphasises that it's generally the interplay between the properties of the numerical algorithm and the problem be solved which together dictate the stability of a particular calculation - what's a good time step (or other numerical parameter) choice for one problem won't necessarily be the case for a different problem. You either need a firm understanding of what is going on, or you should repeat any validation/convergence testing for every different problem or set of parameters you consider!


- By comparing the 1st and 3rd set of results, we can of course recover stability for the 2nd case if we just reduce the time step.

The choice of numerical solver, the stability of the solver, the choice of time step size, and the underlying physics of the problem (i.e. the controlling parameters) are all intimately linked - we will go into this in detail in ACSE-3 where you will gain knowledge and experience in how to choose sensible solvers and time step sizes (and mesh size in the case of PDEs) depending on the problem you're trying to solve - there is no one solver suitable for all problems!

### Solution - the double pendulum

We will use the MSD template from the lecture for this.

````python
from scipy.integrate import solve_ivp

# we should really pass in the parameters rather than hard-coding them
def f_Double_Pendulum(t, state):
    """RHS vector for the double pendulum system

    t: time
    state: state vector [th1, th2, p1, p2]
    """
    # define problems parameters:
    g = 9.8
    m = 1  # assume both masses the same
    L = 1  # assume both lenths the same
    # unpack state
    th1, th2, p1, p2 = state
    # derivatives
    th_diff = th1 - th2
    c_th_diff = np.cos(th_diff)
    s_th_diff = np.sin(th_diff) 
    mL2 = m*L**2
    th1_dot = 6/mL2 * ( 2*p1 - 3*c_th_diff*p2) / ( 16 - 9*c_th_diff**2 )
    th2_dot = 6/mL2 * ( 8*p2 - 3*c_th_diff*p1) / ( 16 - 9*c_th_diff**2 )
    p1_dot = -0.5*mL2 * ( th1_dot*th2_dot*s_th_diff + 3*(g/L)*np.sin(th1))
    p2_dot = -0.5*mL2 * ( -th1_dot*th2_dot*s_th_diff + (g/L)*np.sin(th2))
    return [th1_dot, th2_dot, p1_dot, p2_dot]


# Define initial conditions
th1 = 180*np.pi/180 # convert degree to radians
th2 = 170*np.pi/180 # 180 points straight up
p1 = 0.
p2 = 0.

# Integration window
t0 = 0.
tf = 10.
# output time we want from the sovler
ts = np.linspace(t0, tf, 200)


# A solver tolerance - not really needed here, use default
tol = 1.e-5
# call the solver
sol = solve_ivp(f_Double_Pendulum, [t0, tf], np.array([th1, th2, p1, p2]), method='RK45', t_eval=ts, rtol=tol, atol=tol)

fig, ax1 = plt.subplots(1, 1, figsize=(8, 5))
ax1.plot(ts, sol.y[0,:], 'b', label='x1')
ax1.plot(ts, sol.y[1,:]-sol.y[0,:], 'k', label='x2-x1')
ax1.set_xlabel('$t$', fontsize=16)
ax1.set_ylabel('$\theta(t)$', fontsize=16)
ax1.set_title('Coupled pendulum problem (Angles)', fontsize=16)
ax1.legend(loc='best', fontsize=16)
````

````python
# let's generate a movie
from matplotlib import animation
from IPython.display import HTML

def pendulum_get_xy_from_angles(th1, th2):
    """Convert the two angles describing this problem
    into the x and y locations of the two end points of
    the pendula. Remember that the angles point to the 
    midpoint of the rods making up the pendula
    """
    L = 1.
    x1 = (L)*np.sin(th1)
    y1 = -(L)*np.cos(th1)
    x2 =  x1 + 2*(L*(np.sin(th1) + 0.5*np.sin(th2)) - x1)
    y2 =  y1 + 2*(-L*(np.cos(th1) + 0.5*np.cos(th2)) - y1)
    return x1, y1, x2, y2


# Define initial conditions
th1 = 180*np.pi/180
th2 = 170*np.pi/180
p1 = 0.
p2 = 0.

# Integration window
t0 = 0.
tf = 20.
# output time we want from the sovler
ts = np.linspace(t0, tf, 500)

# A solver tolerance - not really needed here, use default
tol = 1.e-5
# call the solver
sol = solve_ivp(f_Double_Pendulum, [t0, tf], np.array([th1, th2, p1, p2]), 
                method='RK45', t_eval=ts, rtol=tol, atol=tol)

th1 = sol.y[0,:]
th2 = sol.y[1,:]

fig = plt.figure(figsize=(7, 7))
ax1 = fig.add_subplot(111)
ax1.set_xlim(-2.5, 2.5)
ax1.set_ylim(-2.5, 2.5)

# initialise the "lines" making up the moving part of the animation
line0 = ax1.plot([], [], 'k', lw=3)
line1 = ax1.plot([], [], 'k', lw=3)
line2 = ax1.plot([], [], 'ro')
line3 = ax1.plot([], [], 'bo')  
# combine the line objects - these are the things we will update between anin frames
lines = line0 + line1 + line2 + line3
plt.close()

def init():
    for line in lines:
        line.set_data([],[])
    return lines

def data_gen(framenumber, th1, th2):
    x1, y1, x2, y2 = pendulum_get_xy_from_angles(th1[framenumber], th2[framenumber])
    lines[0].set_data([0, x1], [0, y1])
    lines[1].set_data([x1, x2], [y1, y2])
    lines[2].set_data(x1, y1)    
    lines[3].set_data(x2, y2)    
    return lines

anim = animation.FuncAnimation(fig, data_gen, frames=np.arange(0, len(th1)), fargs=(sol.y[0,:], sol.y[1,:]),
                               interval=50, blit=True)

# save the animation as gif or mp4 if needed
#anim.save('animations\double_pendulum.gif', writer='imagemagick')
#anim.save('animations\double_pendulum.mp4', extra_args=['-vcodec', 'libx264'])
# display in notebook
#HTML(anim.to_jshtml())
HTML(anim.to_html5_video())
````
