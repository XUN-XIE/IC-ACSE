## Surrogate Model

There is some overlap between the models described above with the concept of "surrogates", 

or a surrogate might be a thing that is constructed *from* a computational model,  

i.e. it can be interpreted as a lower level in the *conceptual $\rightarrow$ scientific $\rightarrow$ mathematical $\rightarrow$ computational model* hierarchy.

It can also be interpreted in the context of Machine Learning or AI where a "model" is constructed from data directly, not necessarily via the *conceptual  $\rightarrow$ scientific $\rightarrow$ mathematical* ... hierarchy.  

Of course the data might itself come from a computational model so ... these topics are all very jumbled up and overlapping!


[*Wikipedia: Surrogate Model*](https://en.wikipedia.org/wiki/Surrogate_model)

<img src="./figures/surrogate-model-wikipedia.png" width=1000px style="border: 1px solid black">

"A surrogate model is an engineering method used when an outcome of interest cannot be easily directly measured, so a model of the outcome is used instead. Most engineering design problems require experiments and/or simulations to evaluate design objective and constraint functions as a function of design variables. For example, in order to find the optimal airfoil shape for an aircraft wing, an engineer simulates the airflow around the wing for different shape variables (length, curvature, material, ..). **For many real-world problems, however, a single simulation can take many minutes, hours, or even days to complete. As a result, routine tasks such as design optimization, design space exploration, sensitivity analysis and what-if analysis become impossible since they require thousands or even millions of simulation evaluations**.

<br>

One way of alleviating this burden is by constructing approximation models, known as surrogate models ... that mimic the behavior of the simulation model as closely as possible while being computationally cheap(er) to evaluate. **Surrogate models are constructed using a data-driven, bottom-up approach. The exact, inner working of the simulation code is not assumed to be known (or even understood), solely the input-output behavior is important**. A model is constructed based on modeling the response of the simulator to a limited number of intelligently chosen data points. ... When only a single design variable is involved, the process is known as curve fitting.

<br>

Though using surrogate models in lieu of experiments and simulations in engineering design is more common, surrogate modeling may be used in many other areas of science where there are expensive experiments and/or function evaluations."

<br>

"The scientific challenge of surrogate modeling is the generation of a surrogate that is as accurate as possible, using as few simulation evaluations as possible...

<br>

The accuracy of the surrogate depends on the number and location of samples (expensive experiments or simulations) in the design space. Various design of experiments (DOE) techniques cater to different sources of errors, in particular, errors due to noise in the data or errors due to an improper surrogate model."

So you can think of a surrogate model as an approximation of a computational model (which is itself an approximation to the mathematical model, which is an approximation of the real system!).

So there are multiple layers here, this final one is performed since sometimes it's useful for us to have a model that runs virtually instantaneously, on a phone for example, when the computational model may take weeks to run on a supercomputer.

## Discrete vs continuous

Some problems, systems we wish to simulate, etc, are fundamentally discrete and so well-suited to fitting onto (i.e. being solved using) computers which are also discrete. 

However, many problems we wish to solve are continuous.

For example, consider a variable that is a function of space and time - the temperature in this room for example.


### Aside: are space and time actually continuous or discrete? [*]

With a little more thought you may actually question whether space and time are continuous or actually given the fundamental building blocks of matter they are actually discrete

https://www.forbes.com/sites/startswithabang/2018/06/14/are-space-and-time-quantized-maybe-not-says-science/

https://physics.stackexchange.com/questions/9720/does-the-planck-scale-imply-that-spacetime-is-discrete


Well whatever the answer to this is in reality, in practice if these building blocks are very small compared to the scales we are actually interested in, then we make what is called the **continuum or macroscopic assumption** (<https://en.wikipedia.org/wiki/Continuum_mechanics#Concept_of_a_continuum>) and formulate a mathematical model which does assume continuous variables!

So a mathematical modeller may take a discrete system, construct a continuous scientific/mathematical model to describe it (e.g. an ODE or a PDE), and then we as computational scientists need to (re-)discretise it to make it finite/discrete in order to solve it on a computer!!

This may seem crazy, but it does make sense since we can't in general afford to solve a form of the underlying problem that is based on the fundamental discrete objects - we cannot afford to simulate every molecule, particle or atom in this room.

A key job of a numerical modeller/analyst is to decide how to represent a continuous process by something that fits on a computer, i.e. to come up with a discrete model, and to implement it in code.

Often, we represent the "continuous" world with a continuous ODE or PDE. The process of turning this into something we can consider and solve with a computer is called *discretisation*.

In doing this we need to consider many things - this will be covered in detail in the module ACSE3 - Numerical Methods.

To begin though we will keep things simple and start with some simple models that are already discrete.

## Example - the logistic map

The logistic map is a relatively simple discrete relation that can be used to model certain processes.

(NB. It's *nonlinear*, so it's certainly not as simple as it could be, but note its the nonlinearity here that lead to interesting behaviour!)

For example, consider population size and its growth - for a start is this discrete or continuous?

For some biological processes this would be a continuous (or rather assumed to be continuous even though each birth/death is a discrete event!) process - e.g. human population.

However, for others it is best modelled as being discrete - e.g. for species where generations may not overlap or where there are well-defined, discrete breeding seasons.

The logistic map is a famous discrete relation that captures this behaviour. It is given by

$$x_{n+1} = rx_n(1-x_n)$$

where $x_n$ is a value between zero and one ($x_n\in [0,1]$) that represents the ratio of the current population to the maximum possible population.  So $x$ itself can take any value (is continuous, although recall floating point representation of numbers, and see later!!), but $n$ the generation is discrete.

$r$ is a parameter, sometimes termed the *biotic potential* or the *Malthusian parameter*.

When the population size is small ($x_n$ being small implies that $x_n^2$ is very small and so can justifiably be ignored)

$$x_{n+1} = rx_n(1-x_n) = rx_n - rx_n^2  \approx rx_n $$

and so we can think of $r$ as being the growth rate: $r\approx x_{n+1}/x_{n}$ (strictly speaking $r-1$ would generally be defined as the *growth* rate - a growth rate of 10% being equivalent to $r=1.1$).


If our model was simply 

$$x_{n+1} = rx_n$$ 

then we would experience unbounded [exponential growth](https://en.wikipedia.org/wiki/Exponential_growth):

$$ x_{n+1} = rx_n = r(rx_{n-1}) = \ldots = r^{n+1}x_0.$$

What the logistic map looks to do is acknowledge that with finite resources the growth rate will reduce as we approach the maximum population size.  The simplest way to encode this limiting factor mathematically is to make the growth rate depend linearly on the population size, to be a maximum when the population size is small, and to reach zero when we are at maximum population, i.e. replace $r$ with $r \times (1-x)$, where we can interpret $x$ as the population size as a proportion of the maximum possible population - effectively we have [non-dimensionalised](https://en.wikipedia.org/wiki/Nondimensionalization) our problem, see later lectures in this module for more on this. 

Simply by replacing $r$ with $r(1-x_n)$ in the discrete relation directly above is how we end up with the logistic map and non-trivial, interesting dynamics.

Let's write a function that evaluates the right hand side (RHS):

````python
def logistic_map(r, x):
    return r*x*(1-x)
````

As it's easy for us in our IPython environment, let's generate some plots to understand what this does.

````python
# Define a parameter value r
r = 2

# Possible population values (x) are between zero and one,
# create a vector of possible values for plotting purposes
x = np.linspace(0, 1, 100)
# 100 points should be enough to get a nice smooth line, but you 
# could increase this.
# You could also have used np.arange

# Compute the corresponding logistic_map values, as we'll be
# plotting these, let's just use variable y to store these
y = logistic_map(r, x)

# Set up a figure for plotting
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

# plot the data
ax1.plot(x,y,'b')

# add axes labels and a title, play with the fontsizes to make it look nice
ax1.set_title('The logistic map ($r={0:.3f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
ax1.set_ylabel('$rx(1-x)$', fontsize=14)
# add a grid
ax1.grid(True)
````

### Expected behaviour

There are lots of ways we can start thinking conceptually, or theoretically (with a pencil and paper) about what we can do with this simple model, what behaviour we expect, how we might extend it to increase complexity (e.g. to better match a system we want to model) ...

This isn't a module on mathematical modelling or dynamical systems so we won't go into detail, but we can do some fun things to see what sort of avenues we might go down, as well as to practice our coding ...

Firstly, we can consider if this problem has any *fixed points*, i.e. in this example a population that if reached then doesn't vary in future generations. 

(Note that in our model $x=1$ is not a fixed point, this state implies that we will exhaust all our resources in that generation and so in the next discrete step the population goes to zero as everyone dies and no-one is born!)

Mathematically, for our problem, we write this as

\begin{align*}
x_n &= x_{n+1} \\
\iff x_n &= rx_n(1-x_n)
\end{align*}

What two values of $x_n$ does this correspond to?

We can work this our mathematically and/or we can just plot it to figure this out:

````python
r = 3
x = np.linspace(0, 1, 100)
y = logistic_map(r, x)

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

ax1.plot(x,y,'b',label=r'$y = rx(1-x)$')

ax1.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
# add a grid
ax1.grid(True)

ax1.plot(x,x,'k',label=r'$y = x$')
ax1.plot(0, logistic_map(r,0), 'ro')
ax1.plot(1-1./r, logistic_map(r,1-1./r), 'ro', label='intersection points')
ax1.legend(loc='best', fontsize=14)
````

What does this tell us?

Ignoring the fixed point at (0,0) for now, if our population is below the $x$ value that corresponds to the non-zero intersection value it will clearly grow in the next generation. But if it is above that value, the next generation will be lower.

We can plot this evolution of population sizes in what is sometimes referred to as a [cobweb plot](https://en.wikipedia.org/wiki/Cobweb_plot).

````python
def plot_logistic_map_convergence(r, x0, iterations=10, ax=None):
    """ Function to plot the iterations that the logistic map undergoes
    on axes ax (if given)
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, 1, 100)
    y = logistic_map(r, x)
    ax.plot(x,y,'b',label=r'$y = rx(1-x)$')
    ax.plot(x,x,'k',label=r'$y = x$')
    #ax.plot(1-1./r, logistic_map(r,1-1./r), 'ro', label='intersection')  
    ax.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate upto "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = logistic_map(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
````

````python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

r = 1.9
x0 = 0.2
plot_logistic_map_convergence(r, x0, 10, ax1)
````

If it's not clear what this diagram is telling us:

1. We start from an $x$ value - call it $x_0$ and plot it on the $y=x$ line of the plot - it's the green dot (this is the `x = x0` line in the code above).


2. Use the logistic map to find the next value which is given by the blue line - the vertical black dashed line up from the green dot indicates this operation (the `y = logistic_map(r,x)` line in the code above computes the value, and `ax.plot(x, y, ... ` plots it).


3. Now we need to store this new value for use in the next iteration, i.e. we effectively set $x_1 = y$. We don't actually store all the iterations in this implementation, e.g. in a vector of length the number of iterations we are going to perform (we could easily do this), so instead we just overwrite our $x$ variable (`x=y` in the code above) for use in the next iteration. In the diagram this operation is indicated by the horizontal line that takes us back to the $y=x$ curve.


4. And we repeat. To help identify the iterations we change the opacity of the red dots - this will come in more useful shortly!

How would we get the system to converge to the fixed point at zero?

(Think about choice of parameter $r$ rather than a starting location)

There are two parameters we can play with here - the initial condition $x_0$ and the parameter $r$.

Thinking about this as a generic numerical algorithm, it's vitally important that we understand/appreciate the sensitivity to both these parameters:

1. We might not know the starting population with complete accuracy - does this matter?


2. We might not know $r$ either, does this matter? Is $r$ completely knowable or will it always contain inevitable uncertainty? Here we are thinking of $r$ as some "physical" parameter that governs our system.


3. As we will see later $r$ here can also be interpreted as a numerical parameter, i.e. as a user/modeller we need to select an appropriate value - how do we do this and what are the consequences of the wrong choice?

<br>

Let's start by changing the starting value, and also compare what happens if we slightly perturb the starting value:

````python
# Let's have two plots side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

r=1.9
x0 = 0.7
plot_logistic_map_convergence(r, x0, 10, axs[0])
pert = 0.05
plot_logistic_map_convergence(r, x0 + pert, 10, axs[1])
````

So, for the $r$ parameter value considered at least, we always end up at the same place.

For a small perturbation size the subsequent populations sizes seem close - any differences clearly reduce to zero as we ultimately reach the same place, but do the differences decay monotonically?

This is essentially telling us that (for this parameter value at least!), our system dynamics are not sensitive to the initial condition(s).

This could be quite important as in general we won't know the complete initial state of a system we are modelling.

For more on this topic, if interested, read up on chaos: <https://en.wikipedia.org/wiki/Chaos_theory>

### Bifurcation diagram

So if we vary $r$ we can get quite different behaviour - both the intermediate steps as well as where we end up in the long run. 

Let's plot the ultimate behaviour as a function of the parameter $r$ in a so-called [bifurcation diagram](https://en.wikipedia.org/wiki/Bifurcation_diagram).

NB. this implementation is [*vectorised*](https://en.wikipedia.org/wiki/Array_programming). See also <https://www.geeksforgeeks.org/vectorization-in-python/>.  If we looped over all the $r$ values this code would run much slower.

````python
# consider N different r values - the bigger the N we select, the longer it will take!
N = 10000
r = np.linspace(2.0, 4.0, N)
# corresponding N initial x values
x = 0.1*np.ones(N)
# we're interested in long term limiting behaviour, so do quite a few iterations of the map 
iterations = 500
# but only plot the final 100 values we hit
values_to_plot = 100
# compute the logistic map dynamics for each r value, 
# and plot the final x values we hit for each r
fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(111)
for it in range(iterations):
    x = logistic_map(r,x)
    if it>(iterations-values_to_plot):
        ax1.plot(r,x,'k,',alpha=0.1)
ax1.set_title('Bifurcation diagram for the logistic map', fontsize=16)
ax1.set_xlabel('$r$', fontsize=14)
ax1.set_ylabel('Eventual Population', fontsize=14)
````

# Taylor series

Hopefully most of you will have seen Taylor series before.

A Taylor series is a method by which we can approximate an arbitrary function by a sum of terms made up of that function and its derivatives at a single point.

This concept underlies many things (especially in computational science) so it's worth us spending some time reviewing it.


## Motivation - a constant approximation

Consider a function of a single independent variable:  $f(x)$

Suppose this function is really really expensive to evaluate. Suppose I have already evaluated it at a single $x$ value, call it $x_0$, i.e. I know the value of $f(x_0)$.

Given only this information, what is the best guess (estimation/approximation) I can make for $f(x)$ for a choice $x\ne x_0$?

Well I can't really do anything better than 

$$  f(x) \approx f(x_0)$$

This may seem a hopeless approximation, but it can be used in some situations. Pretend $f$ is a weather forecast and $x$ is time. This just says that for a forecast of the weather tomorrow use what we see today - this is a real technique called [*persistent forecast*](https://en.wikipedia.org/wiki/Weather_forecasting#Persistence).  

"... This makes persistence a 'hard to beat' method for forecasting longer time periods": <http://ww2010.atmos.uiuc.edu/(Gh)/guides/mtr/fcst/mth/prst.rxml>

Let's consider an example:

````python
def f(x):
    return np.exp(x)
````

and plot our current situation, assuming that $x_0=0$. We call this the expansion point - we are constructing an approximation "around this point".

````python
fig = plt.figure(figsize=(7, 7))
ax1 = plt.subplot(111)

ax1.set_title('A constant approximation', fontsize=16)
ax1.set_xlabel('$x$', fontsize=16)

# define our x for plotting purposes
x = np.linspace(-3., 3., 1000)

# plot exact function
ax1.plot(x, f(x), 'k', lw=3, label='Exact function')

# define our approximation
x0 = 0.0
y = f(x0)*np.ones_like(x)
ax1.plot(x0, f(x0), 'ro', label='Expansion point')

# and plot the constant approximation
ax1.plot(x, y, 'b', lw=2, label='Constant approximation')

ax1.legend(loc='best', fontsize=14)
````

Can I have any confidence in this guess?


1. Yes if it turns out that the function $f$ is actually a constant.


2. Yes, if I can assume that $f$ doesn't vary very much in the vicinity of $x_0$, and if we are interested in $x$ values that are close to $x_0$.


How can I do better?

## Motivation - a linear approximation (a "linearisation")

The obvious improvement we can make is to approximate the true function with a linear approximation rather than a constant. 

This is a common way to write a linear function:

$$ y = mx + c$$

where $m$ is the slope and $c$ is the intercept - the $y$ value the line hits the $x=0$ axis at.

We consider an approximation, or an expansion, (or in this case a "linearisation") "about a point".  

For us the point is $x_0$ and it makes sense for us to make the slope of the linear approximation the same as the derivative of the function at this point.  

So we are now using the value of $f$ at $x_0$ AND the value of the derivative of $f$ at $x_0$:  $m = f'(x_0)$.

This is the formula for the linear line with this slope:

$$ y = f'(x_0) x + c$$

where choice of $c$ moves the linear line up and down. We want it to pass through the function at $x_0$ as well of course, and this allows us to figure out what value $c$ should take.

Actually we generally write the linearisation in the following form

$$ f(x) \approx f(x_0) + (x-x_0)f'(x_0)$$

so when $(x-x_0)$ is zero, we are at the expansion point and we indeed get $f(x_0)$ on the RHS. As we move away from that point $(x-x_0)$ grows and our approximation adds a correction based on the size of the derivative at $x_0$.

Let's plot this. First we need to define the derivative:

````python
# of course for our example f this is trivial:
def fx(x):
    return np.exp(x)
````

````python
fig = plt.figure(figsize=(7, 7))
ax1 = plt.subplot(111)

ax1.set_title('A constant approximation', fontsize=16)
ax1.set_xlabel('$x$', fontsize=16)

# define our x for plotting purposes
x = np.linspace(-3., 3., 1000)

# plot exact function
ax1.plot(x, f(x), 'k', lw=3, label='Exact function')

# define our approximation
x0 = 0.0
y = (f(x0) + (x-x0)*fx(x0))*np.ones_like(x)
ax1.plot(x0, f(x0), 'ro', label='Expansion point')

# and plot the constant approximation
ax1.plot(x, y, 'b', lw=2, label='Linear approximation')

ax1.legend(loc='best', fontsize=14)
````

This is clearly an improved approximation compared to the constant one.

However the same accuracy points hold: if we are a long way from the expansion point, or if $f$ is very complex, this may not be a good approximation.

Can we do better - yes we can consider the curvature of the black line in addition to its value and its slope, i.e. we can include a term proportional to $f''(x_0)$ to our approximation.

and so on...

This is the formula for the infinite Taylor series *about (or around) the point $x_0$* 

$$f(x) = f(x_0) + (x - x_0) f'(x_0) + \frac{(x - x_0)^2}{2!} f''(x_0) + \frac{(x - x_0)^3}{3!} f'''(x_0) +
\frac{(x - x_0)^4}{4!} f^{(iv)}(x_0) + \ldots$$

An equivalent way of writing this expansion is

$$ f(x_0+h)  = f(x_0) + hf'(x_0) + \frac{h^2}{2!}f''(x_0) + \frac{h^3}{3!}f'''(x_0) + \ldots $$

or replace $h$ by the notation $\Delta x$ or $\delta x$.

As we include more and more terms our approximation improves - see the following animated gif from Wikipedia which explains the point.

#### Taylor series example

More terms equate with a better approximation valid a larger distance from $x_0$.

### Aside: Big O notation [* important, but in the interests of time read through on your own]

When talking about terms in infinite series (terms we will often be forces to truncate, i.e. throw away), or talking about errors, convergence, complexity, run times etc, so-called [Big-O](https://en.wikipedia.org/wiki/Big_O_notation) is very useful.

As an alternative to writing "$ \ldots $" in the above infinite expansions, we can also write 

$$f(x) = f(x_0) + (x - x_0) f'(x_0) + \frac{(x - x_0)^2}{2!} f''(x_0) + \frac{(x - x_0)^3}{3!} f'''(x_0) + \mathcal{O}((x - x_0)^4)$$

or

$$ f(x_0+h) = f(x_0) + hf'(x_0) + \frac{h^2}{2!}f''(x_0) + \frac{h^3}{3!}f'''(x_0) + \mathcal{O}(h^4) $$


What does this mean?

#### In Mathematics

Well simply $\mathcal{O}$ here means "order".

More formally it is used to signify (place bounds on) the limiting behaviour of the magnitude of a mathematical term (or an algorithm's runtime).

So based on the the example above we have simply replaced the terms 

$$ \frac{h^4}{4!} f^{(iv)}(x_0) +  \frac{h^5}{5!} f^{(v)}(x_0) + \ldots $$

with 

$$ \mathcal{O}(h^4) $$

"in the limit" (and implicitly for this use application we are interested in the limit as $h\rightarrow 0$)

i.e. we have written that

$$ \frac{h^4}{4!} f^{(iv)}(x_0) +  \frac{h^5}{5!} f^{(v)}(x_0) + \ldots  =  \mathcal{O}(h^4) $$

this is stating that as $h$ tends to zero the LHS can bounded in magnitude by a term of the form

$$ C h^4 $$

where $C$ is a constant, i.e. there exists a constant $C$ such that for all sufficiently small $h$

$$\left| \frac{h^4}{4!} f^{(iv)}(x_0) +  \frac{h^5}{5!} f^{(v)}(x_0) + \ldots \right| \le C h^4 $$

In our case the point is that, assuming $f$ is a well-behaved function, i.e. it's derivatives are bounded, then there will be a small enough $h$ such that the terms dependent on the 5th and higher powers or $h$ are very small relative to the first term 

They can't be ignored, but we can select a $C$ which is a bit larger than original factor ${h^4}/{4!}$ so that the above holds.

The point here isn't to actually find $C$, this is just notation to help convey a point.

It's important when analysing errors in algorithms.

Note that for this case we were considering the limit of small $h$ and so the lowest power eventually becomes dominant, the opposite situation would occur if we were considering the limit of large $h$.

With errors we are generally thinking about the former case, with run-times (where $h$ might be replaced with some measure of the size of the problem) we are in the latter case.

#### Example

Consider

$$g(y) = 3y^2 - y^3 + 9y^4$$

in the limit as $y\rightarrow 0$.

We can show that

$$\left| 3y^2 - y^3 + 9y^4 \right| \le \left| 3y^2\right| + \left|y^3\right| + \left|9y^4 \right| \le 3y^2 + y^2 + 9y^2 \le 13 y^2$$

as we are considering the case as $y$ gets small. Hence we can write 

$$g(y) = 3y^2 - y^3 + 9y^4 = \mathcal{O}(y^2) \quad\text{as}\quad y\rightarrow 0$$


If we were considering the case of $y\rightarrow \infty$, then we would write instead

$$g(y) = \mathcal{O}(y^4) \quad\text{as}\quad y\rightarrow \infty$$


When dealing with numerical errors you may see things like

$$\text{error} \approx 10^{-5}\Delta t +  10^{5}\Delta t^2$$

For larger values of $\Delta t$ the second term will clearly dominate due to the relative size of the constant in front of it compared to the first term.

But when performing a convergence analysis, i.e. investigating how the error drops as $\Delta t$ is reduced, there will come a point for small enough $\Delta t$ that the first term starts to dominate, and from the point onwards the error will decay linearly rather than quadratically, i.e. halving $\Delta t$ leads to a reduction in the error by a factor of 2 rather than 4.

So we would say that 

$$\text{error} = \mathcal{O}(\Delta t)$$

even though at larger $\Delta t$ values the error would be observed to decay quadratically as we reduce $\Delta t$.

We use the expression "in the asymptotic limit" to refer to region of parameter space where the leading order behaviour dominates. For the above example if we observe something that looks like second order behaviour (when we expect first) we would explain this away by saying that we are not in the asymptotic limit. If we see very close to first order behaviour then we would say we are in the asymptotic limit.

#### In Computer/Computational Science

For the cost of algorithms we use the notation in a similar manner.

An algorithm is said to have (*time* or *algorithmic* or *computational*) [complexity](https://en.wikipedia.org/wiki/Time_complexity) of $\mathcal{O}(n^2)$ for example,
if for large enough $n$, where $n$ is a measure of the problem size, the computational cost grows quadratically - for every doubling of the problem size the cost grows by a factor 4.

When it comes to complexity it's quite common to see things like $\mathcal{O}(n \log n)$, i.e. the algorithm scales worse than linearly, but not as bad as quadratically.

For some examples see <https://en.wikipedia.org/wiki/Computational_complexity_of_mathematical_operations>

# Using Taylor series to solve an ODE numerically

## Deriving the forward Euler scheme

Let's suppose we have an ODE of the form

$$y' = f(t,y) \qquad\left[\text{NB. on the LHS we could also use the notation $\dot{y}$ or $\frac{dy}{dt}$}\right]
$$

This is describing a solution function of the form $y\equiv y(t)$, i.e. the solution $y$ is a function of one variable - $t$.  

$y$ is called the *dependent variable* and $t$ the *dependent variable*.

[NB. in the case where $y$ is multi-dimensional - a vector of unknowns/dependent variables - we would talk about a *system of ODEs* - we'll see this case in lecture 3. For the case where there are more than one independent variables then we need to start thinking about *partial differential equations (PDEs)*, covered later in this module and in ACSE-3].

For example

$$y' = \exp(t), \qquad\text{or equivalently} \qquad y' = y$$

the first is an example where the RHS is a function of $t$ only, the second a function of $y$ only. 

This is an example where both $t$ and $y$ appear on the RHS (we will return to this example in lecture 4):

$$y' = y + t^3$$

The first two ($y'=y$ or $y'=\exp(t)$) are relatively easy for us to solve analytically, the final case is a bit more challenging and we may need to resort to a numerical approach (or attempt symbolic computation - see lecture 4).

We can formulate a simple numerical solver using a truncated Taylor series.

From above we know that (assuming smoothness of the solution) the following is true:

$$ y(t+\Delta t) = y(t) + \Delta t \, y'(t) + \frac{\Delta t^2}{2!}y''(t) + \frac{\Delta t^3}{3!}y'''(t) + \mathcal{O}(\Delta t^4),$$
 

Using the notation $y_n=y(t_n), \; y'_n=y'(t_n)$ etc, and assuming a uniform time step size $\Delta t$, this is equivalent to 

$$ y_{n+1} = y_n + \Delta t\, y'_n + \frac{\Delta t^2}{2!}y''_n + \frac{\Delta t^3}{3!}y'''_n + \mathcal{O}(\Delta t^4).$$

Dropping second-order terms (which we can justify if $y$ is "smooth", which means that $y''$ is "well-behaved", and $\Delta t$ is chosen to be sufficiently small so that the terms containing higher powers of $\Delta t$ are relatively speaking *very small*), 

and noting that $y'_n = f(t_n,y_n)$ we are left with the discrete relation

$$y_{n+1} = y_n + \Delta t \, f(t_n,y_n).$$

<br>

This is a very famous ODE solver, or time-stepping method - termed the **forward Euler** or **explicit Euler** method.

We have turned the continuous problem into a discrete one - we have discretised. And note that this is in a form similar to our logistic map, so we know that we can write a simple piece of code that iterates over $n$ to evaluate/predict/approximate the dynamics.

Let's consider the simple model problem 

$$y' = y$$

with the initial condition

$$y(0)=1$$

and see what one step of the forward Euler method does, we will assume that we start at some $t$ from the exact solution at that time, i.e. the starting point indicated by the red dot in the following image is on the exact solution (black) line.

````python
fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)
ax1.set_title('One step of Forward Euler', fontsize=16)
ax1.set_xlabel('$t$', fontsize=16)
ax1.set_ylabel('$y(t)$', fontsize=16)

t = np.linspace(0, 1, 1000)

# the RHS of the ODE
def f(t,y):
    return y

# an example solution trajectory (or solution to ODE $\dot{y}=f(y), y(0)=1$)
def y(t):
    return np.exp(t)

ax1.plot(t, y(t), 'k', label='Trajectory $y=y(t)$')

# this is just some location in time
t0 = 0.35
# and an example of a time step size large enough we can see what the method is doing
dt = 0.4

ax1.plot([t0], [y(t0)], 'ro')
ax1.plot([t0, t0 + dt], [y(t0), y(t0) + dt * f(t0, y(t0))], 'b', label = r"$y'=f(t_0,y_0)$")

ax1.plot([t0 + dt], [y(t0 + dt)], 'ko', label='Exact soln at $t_0 + \Delta t$')
ax1.plot([t0 + dt], [y(t0) + dt * f(t0, y(t0))], 'go', label='FE numerical soln at $t_0 + \Delta t$')
ax1.legend(loc='best', fontsize=14);
````

The black dot in the figure above shows what the true/exact solution at the new $n$ level is, while the green dot shows what the numerical solver predicts it to be.

The difference between the two is the error that has been introduced in a single step of the method.

Now in the next step of the algorithm we of course have to start from the green dot as we assume in practice we don't know the black line.  

The next step computes

$$y_{n+1} = y_n + \Delta t \, f(t_n,y_n),$$

but note that $f$ for this particular simple test case is simply $y$. 

We assume that we don't know what the exact solution is at this time level, and hence we can't read off the correct $t$ from the $x$ axis in the above plot and compute $\exp(t)$ in the RHS of our update. If we could then this would let us use the correct new slope at this time, i.e. the slope at the black dot.  

Instead we are forced to read off the $y$ axis - notice that in this example this gives us a $y$ value that is too small, and hence an $f$ that is too small. Therefore our next update will give us a second point that is even more too small!

The moral of the story here is that **errors generally accumulate**, as we can see if we plot this next step of the algorithm:

````python
fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)
ax1.set_title('Two steps of Forward Euler', fontsize=16)
ax1.set_xlabel('$t$', fontsize=16)
ax1.set_ylabel('$y(t)$', fontsize=16)

t = np.linspace(0, 1.5, 1000)

# the RHS of the ODE
def f(t,y):
    return y

# an example solution trajectory (or solution to ODE $\dot{y}=f(y), y(0)=1$)
def y(t):
    return np.exp(t)


ax1.plot(t, y(t), 'k', label='Trajectory $y=y(t)$')

# this is just some location in time
t0 = 0.35
# and an example of a time step size large enough we can see what the method is doing
dt = 0.4


ax1.plot([t0], [y(t0)], 'ro')
ax1.plot([t0, t0 + dt], [y(t0), y(t0) + dt * f(t0, y(t0))], 'b')
y1 = y(t0) + dt * f(t0, y(t0))
ax1.plot([t0 + dt, t0 + 2*dt], [y1, y1 + dt * f(t0+dt, y1)], 'b')

ax1.plot([t0 + dt, t0+2*dt], [y(t0 + dt), y(t0+2*dt)], 'ko')
ax1.plot([t0 + dt, t0 + 2*dt], [y(t0) + dt * f(t0, y(t0)), y1 + dt * f(t0+dt, y1)], 'go')
````

## A continuous model (an ordinary differential equation)

Let's start with an ODE of the form

\begin{align*}
\frac{dx}{dt} = \dot{x} = x' = \alpha x(1-x)
\end{align*}

with $\alpha$ some parameter.

### Finding an analytical solution ("pen and paper") [*]

We can use a technique called *separation of variables* to find an analytical solution

\begin{align*}
\frac{dx}{dt} &= \alpha x(1-x) \\[10pt]
\implies \int \frac{dx}{x(1-x)} &= \int \alpha\, dt\\[10pt]
&\left[ \text{We can use the fact that   } \frac{1}{x(1-x)} = \frac{1}{x}+\frac{1}{1-x} \right] \\[10pt]
\implies \int \frac{dx}{x} +  \int \frac{dx}{1-x} &= \int \alpha\, dt\\[10pt]
\implies \log(|x|) -\log(|1-x|) &= \alpha t + C\\[10pt]
\implies \log(|1-x|) - \log(|x|) &= -\alpha t - C\\[10pt]
\implies \log\left(\left| \frac{1-x}{x}\right|\right) &= -\alpha t - C\\[10pt]
\implies \left| \frac{1-x}{x}\right| &= \exp(-\alpha t - C)\\[10pt]
\implies \frac{1-x}{x} &= A\exp(-\alpha t)\\[10pt]
\implies x(t) &= \frac{1}{1+A\exp(-\alpha t)}
\end{align*}

where $C$ is an arbitrary constant of integration, $A=\pm\exp(-C)$ is its exponential, with a sign that we can choose to allow us to drop the absolute value.

If we plug in the initial condition this allows us to fix the constant:
\begin{align*}
x(0)=\frac{1}{1+A} \implies A = \frac{1-x(0)}{x(0)}
\end{align*}

and so we have the solution

$$x(t) = \frac{x(0)}{x(0) + (1-x(0))e^{-\alpha t}}$$

We will often use the notation  $x_0 = x(t=0)$, and so write this as

$$x(t) = \frac{x_0}{x_0 + (1-x_0)e^{-\alpha t}}$$

<br>

Note that we will solve this problem (and hopefully find the same solution) using symbolic computation in a homework exercise in lecture 4.

## Comparing numerical and analytical solutions qualitatively

So now we have an exact solution, and we know how to solve the ODE numerically using the forward Euler scheme. 

Let's code both up and see how well they agree.

In the example below we will see what happens qualitatively if we vary the time step (i.e. now we're not checking sensitivity to any physical parameter as we did above, but here sensitivity to a numerical parameter).

````python
def logistic_rhs(alpha, x):
    return alpha*x*(1-x)
````

````python
def x_ex(t, x0, alpha):
    """ Function to evaluate the exact solution to the logistic problem.
    """ 
    return x0/(x0 + (1-x0)*np.exp(-alpha*t))
````

The key lines that implement the forward Euler solver in the code below are

```Python
for n in range(len(t)-1):
    x[n+1] = x[n] + dt * logistic_rhs(alpha, x[n])
```

the rest is setting up space to store the result and plotting.

````python
# set some physical parameters
alpha = 1.0

# set some numerical (time-stepping) parameters
dt = 1
tend = 10

def compute_and_plot_logistic(dt, tend, alpha, x0, ax):
    # define our time levels with a fixed dt
    t = np.arange(0, tend, dt)
    # initialise an array to store all solution values (for later plotting)
    x = np.empty(len(t))
    # and set the first entry in array to the initial condition
    x[0] = x0

    # now let's do the FE time-stepping via a 'for' loop (of course other approaches possible)
    for n in range(len(t)-1):
        x[n+1] = x[n] + dt * logistic_rhs(alpha, x[n])

    ax.plot(t, x, 'b.-', label='numerical approximation')
    # let's use a finer resolution to plot the exact solution to give a smoother line.
    t_fine = np.arange(0, tend, dt/100)
    ax.plot(t_fine, x_ex(t_fine, x0, alpha), 'k', label='exact solution')
    ax.set_xlabel(r'$t$', fontsize = 16)
    ax.set_ylabel(r'$x$', fontsize = 16)
    ax.set_title('Numerical vs exact solution', fontsize = 16)
    ax.grid(True)
    ax.legend(loc='best', fontsize = 14)


# Let's have a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
# this turns the 2D array of axes labels into a 1D vector we can easily call 
axs = axs.reshape(-1)

# fix the physical parameters and the initial condition
alpha = 0.95
x0 = 0.1
tend = 100

# just vary the time step size
dt = 1
compute_and_plot_logistic(dt, tend, alpha, x0, axs[0])

dt = 2
compute_and_plot_logistic(dt, tend, alpha, x0, axs[1])

dt = 3
compute_and_plot_logistic(dt, tend, alpha, x0, axs[2])

dt = 3.16
compute_and_plot_logistic(dt, tend, alpha, x0, axs[3])
````

Should we have expected this behaviour?

Yes if we do a little bit of analysis of our numerical solver.

Note that for this problem the solver implements:

\begin{align*}
x_{n+1} &= x_n + \Delta t \, f(t_n,x_n) \\
& = x_n + \Delta t  \alpha \, x_n(1-x_n)
\end{align*}

But notice that this is exactly the same as the second form of our discrete logistic map with 

$$r = \alpha \Delta t$$

We know that the behaviour of the discrete logistic map is completely dependent on the choice of parameter $r$. 

Based on this discussion and equivalence we expect our numerical solution to display identical behaviour, but now this is governed by the product $\alpha \Delta t$ - i.e. a combination of a *physical parameter* and a *numerical parameter*.

For example, if $\alpha \Delta t > 3$ we would expect our numerical simulation to blow up - and this is exactly what we see, and the $\Delta t$ value that leads to stability/instability  is completely dependent on an underlying physical parameter - if we have a different $\alpha$ then the $\Delta t$ choice we need to make to get a stable numerical solution changes!

Similarly in the plots above we see that oscillatory or random (but technically stable) behaviour is seen for $\Delta t$ values that completely agree with the analysis we can carry over from our investigation of the discrete logistic map.

## Comparing numerical and analytical solutions quantitatively

The great thing about have an exact solution (we won't always be so lucky, and of course if we were we wouldn't have a need for numerical approximations!) is that we can compute errors quantitatively and see how they vary with our $\Delta t$ choice - we are conducting a "sensitivity study" or a "convergence analysis".

Note to avoid the fact that long term behaviour might converge to a steady state, here we only integrate up to a time we know from above is in the "unsteady" phase of the problem. When performing a convergence analysis we are doing multiple runs, some requiring lots of iterations, and hence limiting the simulation time means we conclude our analysis more rapidly.

````python
def logistic_rhs(alpha, x):
    return alpha*x*(1-x)

def x_ex(t, x0, alpha):
    """ Function to evaluate the exact solution to the logistic problem.
    """ 
    return x0/(x0 + (1-x0)*np.exp(-alpha*t))
    
def time_step(alpha, x0, t0, t_max, dt):
    """ Implement simple time-stepper
    """
    x = np.array(x0)
    t = np.array(t0)
    x_all = [x0]
    t_all = [t0]  
    while t < t_max:
        # start from the most recent value
        x = x_all[-1]
        x = x + dt * logistic_rhs(alpha, x)
        x_all.append(x) 
        t = t + dt
        t_all.append(t)
    return np.array(x_all), np.array(t_all)

alpha = 0.95
t0 = 0.0
x0 = 0.1

dt = 5.
# as this problem approaches an attractor at large times - this obfuscates the early time errors
# so only run for a relatively short time
t_max = 5.
errors = []
dt_array = []
while dt > 1.e-4:
    dt_array.append(dt)
    x_all, t_all = time_step(alpha, x0, t0, t_max, dt)
    errors.append(abs(x_all[-1] - x_ex(t_max,x0,alpha)))
    dt *= 0.5

fig = plt.figure(figsize=(16, 5))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

# plot error vs dt normally
ax1.plot(dt_array, errors, 'k.', label='Error with forward Euler')
ax1.set_xlabel('$\Delta t$', fontsize=14)
ax1.set_ylabel('Error', fontsize=14)
ax1.set_title('Forward Euler convergence plot (end val)', fontsize=16)
ax1.grid(True)


# plot error vs dt with a log-log plot
ax2.loglog(dt_array, errors, 'k.', label='Error with forward Euler')
ax2.set_xlabel('$\Delta t$', fontsize=14)
ax2.set_ylabel('Error', fontsize=14)
ax2.set_title('Forward Euler convergence plot (RMS)', fontsize=16)
ax2.grid(True)

# use numpy.polyfit to find best fit linear line to data
# we expect linear asymptotic behaviour - use this parameter to discount some of  the larger dt values
start_fit = 3

line_fit1 = np.polyfit(np.log(dt_array[start_fit:]), np.log(errors[start_fit:]), 1)
ax2.loglog(dt_array[start_fit:], np.exp(line_fit1[1]) * dt_array[start_fit:]**(line_fit1[0]), 'k-', label = 'slope: {:.2f}'.format(line_fit1[0]))

# add a nice way to indicate slope
annotation.slope_marker((5.e-3, 2e-5), (1, 1), ax=ax2,
                        size_frac=0.25, pad_frac=0.05, text_kwargs = dict(fontsize = 14))

ax1.legend(loc='best', fontsize=14)
ax2.legend(loc='best', fontsize=14)
````

Some crazy things go on for too large a time step size - we see this qualitatively in the figures further above, as well as in these error plots -  for the `dt` values to the right in the plot above. Due to this we discard the first few values when computing the slope of the line fit (`start_fit = 3`).

The slope of the line in the log-log plot if we discard these values is very close to 1.

We will see in later modules that this is expected behaviour - the error convergence rate we expect from theory (here 1 for the forward Euler ODE solver) is only expected to hold in the "asymptotic limit".

So for smaller and smaller $\Delta t$ values we expect to get closer to a slope of 1 - halving $\Delta t$  leads to exactly half the error.

# Summary

So what have we learnt?


1. The differing, over-lapping roles of the "physicist", the mathematical modeller and the computational scientist.



2. Modelling involves multiple steps - it takes us from some real process to a solution obtained on a finite computer. Assumptions and approximations are utilised at every stage. Choices made at every stage could be the difference between a **useful** model (and numerical result) and a useless one.



3. Simple problems can have very complex behaviour.



4. Physical as well as numerical parameter choices (and their combination) may or may not lead to wildly different solutions, and hence both lead to useful or useless results.



5. This last point is related to the concept of **stability** - an algorithm can go unstable which means its solution is useless. Often but not always this involves the algorithm returning an essentially infinite value - informally we say the simulation/calculation has *blown-up*. 



6. So stability is generally dependent on a combination of both numerical parameter choices as well as the underlying "physics".  If you change the "physics" you will generally need to change the numerical parameter choices (or even the entire solver choice) to get good results.


7. Errors accumulate (and how we can use big-O notation to both write down the sizes of errors as well as the complexity of algorithms).
