$$
\begin{align}
\frac{dv}{dt} =& \frac{-C_D\rho_aAv^2}{2m} + gsin\theta \\
\frac{dm}{dt} =& \frac{-C_H\rho_aAv^3}{2Q} \\
\frac{d\theta}{dt} =& \frac{gcos\theta}{v}-\frac{C_L\rho_aAv}{2m}-\frac{vcos\theta}{R_p+z} \\
\frac{dz}{dt} =& -vsin\theta \\
\frac{dx}{dt} =& \frac{vcos\theta}{1+z/R_p} \\
\frac{dr}{dt} =& [\frac{7}{2}\alpha \frac{\rho_a}{\rho_m}]^\frac{1}{2}v \\
\end{align} \\
where\ A=\pi r^2,\ C_D=1,\ C_H=0.1,\ C_L=10^{-3},\ R_p=6371000,\ H=8000,\ \alpha=0.3, Q=10^7,\ \rho_m=3000\\
\rho_a = \rho_0 e^{-z/H}
$$

```python
import numpy as np
from scipy.integrate import odeint


def dmove(Point,t,sets):
    C_D, g, C_H, Q, C_L, R_p, alpha, rho_m, rho_0, H, Y = sets
    v, m, theta, z, x, r = Point
    A = np.pi*(r**2)
    rho_a = rho_0*np.exp(-z/H)
    return np.array([(-C_D*rho_a*A*(v**2))/(2*m) + g*np.sin(theta),
                     (-C_H*rho_a*A*(v**3))/(2*Q),
                     (g*np.cos(theta))/(v) - (C_L*rho_a*A*v)/(2*m) - (v*np.cos(theta))/(R_p + z),
                     -v*np.sin(theta),
                     (v*np.cos(theta))/(1 + z/R_p),
                     v*((7/2*alpha*rho_a/rho_m)**(1/2)) if rho_a * (v**2) >= Y else 0])

tmax = 1000
t = np.arange(0,tmax,0.05)
init_state = (21e3,4000*np.pi*(10**3),np.pi/4,100e3,0.,10.)
P1 = odeint(dmove,init_state,t,args = ([1.,9.81,0.1,1e7,1e-3,6371000.,0.3,3000,1.2,8000.,1e5],))
energy_t = -(0.5*P1[1:, 1]*(P1[1:, 0]**2) - 0.5*P1[:-1, 1]*(P1[:-1, 0]**2))/(t[1:]-t[:-1])
energy_z = (0.5*P1[1:, 1]*(P1[1:, 0]**2) - 0.5*P1[:-1, 1]*(P1[:-1, 0]**2))/(P1[1:, 3]-P1[:-1, 3])

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,12))
# ax = Axes3D(fig)
plt.subplot(331)
plt.plot(t,P1[:, 0]) # v-t
plt.title("v-t")
plt.subplot(332)
plt.plot(t,P1[:, 1]) # m-t
plt.title("m-t")
plt.subplot(333)
plt.plot(t,P1[:, 2]*180/np.pi) # theta-t
plt.title("$theta-t$")
plt.subplot(334)
plt.plot(t,P1[:, 3]) # z-t
plt.axhline(y=0,ls=":",c="r")
plt.title("z-t")
plt.subplot(335)
plt.plot(t,P1[:, 4]) # x-t
plt.title("x-t")
plt.subplot(336)
plt.plot(t,P1[:, 5]) # r-t
plt.title("r-t")
plt.subplot(337)
plt.plot(P1[:, 3],P1[:,0]) # v-z
plt.title("v-z")
plt.subplot(338)
plt.plot(energy_z/(4.184*1E12), P1[1:, 3]) # Energy-z
plt.title("Energy-z")
plt.subplot(339)
plt.plot(P1[:, 4],P1[:, 3]) # z-x
plt.title("Trace")
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp


def dmove(t, Point):
    C_D, g, C_H, Q, C_L, R_p, alpha, rho_m, rho_0, H, Y = (1.,9.81,0.1,1e7,1e-3,6371000.,0.3,3000,1.2,8000.,1e5)
    v, m, theta, z, x, r = Point
    A = np.pi*(r**2)
    rho_a = rho_0*np.exp(-z/H)
    return np.array([(-C_D*rho_a*A*(v**2))/(2*m) + g*np.sin(theta),
                     (-C_H*rho_a*A*(v**3))/(2*Q),
                     (g*np.cos(theta))/(v) - (C_L*rho_a*A*v)/(2*m) - (v*np.cos(theta))/(R_p + z),
                     -v*np.sin(theta),
                     (v*np.cos(theta))/(1 + z/R_p),
                     v*((7/2*alpha*rho_a/rho_m)**(1/2)) if rho_a * (v**2) >= Y else 0])

tmax = 3
t = np.arange(0,tmax,0.05)
initial_state = np.array([21000, 3000*4/3*np.pi*10**3, np.pi/4, 100e3, 0, 10])
tol = 1e-4
P1 = solve_ivp(dmove, (0, tmax), init_state, t_eval=np.arange(0,tmax,0.05), method='RK45', atol=tol, rtol=tol)
print(len(P1['y'][0]), len(P1['y'][1]), len(P1['y'][2]), len(P1['y'][3]), len(P1['y'][4]), len(P1['y'][5]), len(t))
energy_z = (0.5*P1['y'][1, 1:]*(P1['y'][0, 1:]**2) - 0.5*P1['y'][1, :-1]*(P1['y'][0, :-1]**2))/(P1['y'][3, 1:]-P1['y'][3, :-1])

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,12))
# ax = Axes3D(fig)
plt.subplot(331)
plt.plot(t,P1['y'][0]) # v-t
plt.title("v-t")
plt.subplot(332)
plt.plot(t,P1['y'][1]) # m-t
plt.title("m-t")
plt.subplot(333)
plt.plot(t,P1['y'][2]*180/np.pi) # theta-t
plt.title("$theta-t$")
plt.subplot(334)
plt.plot(t,P1['y'][3]) # z-t
plt.axhline(y=0,ls=":",c="r")
plt.title("z-t")
plt.subplot(335)
plt.plot(t,P1['y'][4]) # x-t
plt.title("x-t")
plt.subplot(336)
plt.plot(t,P1['y'][5]) # r-t
plt.title("r-t")
plt.subplot(337)
plt.plot(P1['y'][3],P1['y'][0]) # v-z
plt.title("v-z")
plt.subplot(338)
plt.plot(energy_z/(4.184*1E12), P1['y'][3, 1:]) # Energy-z
plt.title("Energy-z")
plt.subplot(339)
plt.plot(P1['y'][4],P1['y'][3]) # z-x
plt.title("Trace")
plt.show()
```
