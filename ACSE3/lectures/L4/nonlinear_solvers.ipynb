# Nonlinear Equations

A **system is said to be linear** if it can be represented by a **linear function** (*i.e.* a straight line in 1D):


$$ f(x) := a + kx, $$


where $f(x)$ is a quantity of interest that depends on some variable $x$, which can be quantified through measurements or observations.

To a linear system we can associate a **linear equation**, which can be written in the form:


$$ f(x) = 0 \;\;\;\; \text{where} \;\;\;\; f(x) := a + kx. $$


We've set the *function* equal to something (here we chose zero) to obtain an *equation*.

Our aim is to find a solution $x$ to this equation.

## Aside 1 (link to the linear systems seen in lecture 3)

Note that the linear (matrix) systems ($A \boldsymbol{x} = \boldsymbol{b}$) from the previous lecture are linear. 

In this case we can define


$$\boldsymbol{f}(\boldsymbol{x}) := A \boldsymbol{x} - \boldsymbol{b}, $$


where $\,\boldsymbol{f}\,$ is a vector-valued function of a vector argument $\boldsymbol{x}$. 

Note first that solving this system for $\boldsymbol{x}$ is completely equivalent to the problem we considered in the previous lecture, and second that linear systems can of course be considered as a subset of all nonlinear problems, and the methods from the previous lecture are solvers appropriate for this subset.

One way to make a matrix system such as this no longer linear (i.e. nonlinear), would be to make the matrix a function of the unknown $x$:

$$ A(\boldsymbol{x}) \boldsymbol{x} = \boldsymbol{b}.$$

For most of this lecture we will consider the scalar case, but will consider the solution of nonlinear *systems* at the end of this lecture and as a homework exercise.

## Aside 2 (a warning on terminology)

Note that we need to be a little bit careful in our use of language here, and there is scope for confusion!  

The above are indeed linear functions/equation/systems, but there is also the concept of a [*linear map, transformation or operator*](https://en.wikipedia.org/wiki/Linear_map) for which the following definition applies:


$$\boldsymbol{f}(c_1\boldsymbol{x}_1 + c_2\boldsymbol{x}_2) = c_1\boldsymbol{f}(\boldsymbol{x}_1) + c_2\boldsymbol{f}(\boldsymbol{x}_2),$$


for scalars $c_1$ and $c_2$.

Notice that a non-zero $a$ or $\boldsymbol{b}$ in the linear functions/equations/systems we defined above means that this particular definition of linearity does not hold:

$$ f(x_1 + x_2) = a + k(x_1 + x_2) =  a + kx_1 + kx_2 \ne a + kx_1 + a + kx_2 = f(x_1) + f(x_2). $$

In areas where this definition of linearity applies, a *linear* function/map/... with the *addition of a constant* is termed an *affine* function/map/...  

Said another way, in this context linear maps would map the origin to the origin (the intercept in zero).

Search for affine here for more information on this distinction in definitions
[https://en.wikipedia.org/wiki/Linear_function](https://en.wikipedia.org/wiki/Linear_function) and [https://en.wikipedia.org/wiki/Linear_equation](https://en.wikipedia.org/wiki/Linear_equation)

In this lecture we will use the definition of linear we gave in the previous cell (i.e. allow the addition of a constant).

## Examples of nonlinearity

Note that the majority of real world systems cannot be expressed in such simple (linear) terms. 

[Note also that discretisations of nonlinear ODEs or PDEs (e.g. being nonlinear in order to accurately represent the nonlinear real world) results in nonlinear discrete systems for which we need to make use of techniques such as those introduced in this lecture. We will see an example of this when we deal with Computational Fluid Dynamics.]

They are then called nonlinear systems, and equations representing them contain terms like:


-  $ x^2,\; x^3,\ldots\;\;\;$ e.g. higher ($>1$) order powers that would appear in polynomials.


- $ \sqrt{x},\; x^{1/3}, x^{\pi},\ldots \;\;\; $ e.g. roots, radicals and other non-integer polynomials.


- $\tan(x),\; \sin(x), \; \log(x),\; \text{erf}(x), \ldots \;\;\; $ e.g. trigonometric and other [*special functions*](https://en.wikipedia.org/wiki/Special_functions).


You should be familiar, for example, with quadratic equations $f(x) := a x^2 + bx + c = 0$, for which there exist the well known analytic solutions:

$$
x^*_{1,2} := \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}\,.
$$

The solutions $\,x^*_{1,2}\,$ are commonly referred to as the **roots** of the equation - this means that $\,f(x^*_{1}) = f(x^*_{2}) =0$.

## An example nonlinear equation

An example of a nonlinear problem without trivial solution is:

$$ x = \tan(x). $$

There is no exact analytic solution to this equation, so we have to devise methods to find approximate solutions of this equation.

That is, we are looking for  $x_{\text{approx}}$ such that 

$$x _{\text{approx}} \approx \tan(x_{\text{approx}}), $$ 

to a set *tolerance* (or *precision*). 

Note that we can of course equivalently look to solve the problem 

$$ f(x) \approx 0 \;\;\;\;\text{where}\;\;\;\; f(x):= x - \tan(x), $$  

[or of course equivalently we could instead choose to define $\,f(x):= \tan(x) - x$].

We will often do this (move all terms of our equation to one side and set equal to zero) to make our presentation of the theory below, as well as our implementation of solvers, generic. In this context a solution to the problem $f(x)=0$ is called a **root**.

# Picard's method of successive approximations (or fixed point iteration)


## Fixed points

Before moving on to problems written in the completely general form $\,f(x) = 0\,$, in this section we will consider equations that can be written as:


$$ x = g(x), $$


where $g$ is a given function.

[Of course if we're given a problem of the second type we can make it look like the first simply by defining $\,f(x) := x-g(x)\,$ or $\,f(x) = g(x)-x,\,$ as we did for the problem involving $\,\tan\,$ above]

An $x$ that satisfies this equation is sometimes called a **fixed point** of the function $g$.

It's called this since if we keep applying the function $g$ to the output, then the output never changes, i.e.


$$x = g(x) = g(g(x)) = g(g(g(x))) = \ldots $$


As an example, consider the function


$$ g(x) := x^{2}-3x+4. $$


This has the fixed point 2, since $g(2)=2$.

While the example


$$ g(x) = x+1, $$


clearly has no fixed points - every application of the function adds another one to our $x$.

If we plot the function $y = g(x)$ as a curve in $(x,y)$ space then the fixed points (if they exist) are the locations where this curve intersects with the line $y=x$.

Let's see some examples.

````python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.tight_layout(w_pad=4)

x = np.linspace(-1,4,100)
ax1.plot(x, x**2 - 3*x + 4, 'b', label='$y = g(x) = x^2-3x+4$')
ax1.plot(x, x, 'k', label='$y = x$')
ax1.plot(2, 2, 'ro', label='intersection')
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel('$y$', fontsize=16)
ax1.set_title('Fixed point example 1', fontsize=16)
ax1.legend(loc='best', fontsize=14)

x = np.linspace(-1,4,100)
ax2.plot(x, x+1, 'b', label='$y = g(x)=x+1$')
ax2.plot(x, x, 'k', label='$y = x$')
ax2.set_xlabel('$x$', fontsize=16)
ax2.set_ylabel('$y$', fontsize=16)
ax2.set_title('Fixed point example 2', fontsize=16)
ax2.legend(loc='best', fontsize=14)

x = np.linspace(-1,10,1000)
y = np.tan(x)
# don't plot large values - comment out next two lines to see issue
y[y>100.] = np.inf
y[y<-100.] = -np.inf
ax3.plot(x, y, 'b', label=r'$y = g(x)=\tan(x)$')
ax3.plot(x, x, 'k', label='$y = x$')
# I know that tan(0) = 0, so I already know one of the roots 
ax3.plot(0, 0, 'ro', label='intersection')
# NB. the following values found below using bisection
ax3.plot(4.493409156799316, 4.493409156799316, 'ro')
ax3.plot(7.725251197814941, 7.725251197814941, 'ro')
ax3.set_ylim([-10,10])
ax3.set_xlabel('$x$', fontsize=16)
ax3.set_ylabel('$y$', fontsize=16)
ax3.set_title('Fixed point example 3', fontsize=16)
ax3.legend(loc='best', fontsize=14)
````

## Algorithm (Picard iteration)

The simplest method to find approximate fixed points/roots to this kind of equation consists in guessing $x$, plugging it into the right-hand-side $g(\cdot)$ and evaluating the function. 

The resulting new value for $x$ is then used as an updated guess. 

A sequence of $x^{(k)}$ values are thus defined via the iteration:

$$ x^{(k+1)} = g\left(x^{(k)}\right), $$

starting from some initial guess $x^{(0)}$.

**Aside:** at this point you might find it helpful to recall our simple iterative solver from the previous lecture:

$$\boldsymbol{x}^{(k+1)} = N^{-1} \left( P \boldsymbol{x}^{(k)} + {\boldsymbol{b}}\right), \;\;\;\; k = 0, 1, \dots$$


In the section on "General framework for stationary iterative methods" we established that we expect the iteration to converge if all the eigenvalues of $M:=N^{-1}  P$ are less than one in magnitude.

This procedure can be repeated until some stopping criteria is met, e.g. when two consecutive values for $x$ differ by less than some user-defined tolerance.   

Note that here 

$$x^{(k+1)} - x^{(k)} = g\left(x^{(k)}\right) - x^{(k)}$$ 

and so the 'change in $x$' value dropping below our tolerance is equivalent to us satisfying $x=g(x)$ to that same tolerance!

This strategy is often referred to as (Picard's) method of successive approximations. 

A pseudo-code description of this solution strategy looks like

```
guess x
x_previous := x + 2*tolerance   # so that the initial evaluation in the while loop is true.
while ( abs(x - x_previous) > tolerance ) do   
    x_previous := x
    x := g(x_previous)
```

## Implementation 

Let's implement a Picard iteration and apply it to solve the problem 

$$x=\mathrm{e}^{-x}$$

````python
def picard(f, x, atol=1.0e-6):
    """ Function implementing Picard's method.
    
    f here is the function g(.) described in the lecture and we are solving x = g(x).
    
    x is an initial guess.
    
    atol is a user-defined (absolute) error tolerance.
    """
    # record number of function evaluations so we can later compare methods
    fevals = 0
    # initialise the previous x simply so that while loop argument is initially true
    x_prev = x + 2*atol
    while abs(x - x_prev) > atol:
        x_prev = x
        x = f(x_prev) # one function evaluation
        fevals += 1
        print('Current iteration solution: ',x)
    print('\nPicard used', fevals, 'function evaluations')
    return x



def g(x):
    return np.exp(-x)


print('\nSolution from Picard: ', picard(g, 0.9, atol=1.0e-7))  # 0.9 is our initial guess


# let's check our answer against a SciPy function: sop.newton.
def f(x):
    return x - np.exp(-x)

print('\nCheck this against the solution from SciPy: sop.newton(f, 0.9)=', sop.newton(f, 0.9))

# NB. if we tighten up the atol toelrance in our call to picard the answer gets closer to SciPy 
# but of course takes more iterations.
````

To get a better idea of what is going on here let's look at the progression of the intermediate results $\,x^{(k+1)}\,$ from $\,x^{(k+1)} = g\left(x^{(k)}\right)\,$, and plot them on top of the curve $y = \mathrm{e}^{-x}$.

````python
def plot_picard_convergence(f, x, ax, max_labels=6, atol=1.0e-4, flabel=''):
    """ Function to demonstrate/plot the Picard algorithm and its convergence
    """
    # variables to store solution iterations for plotting purposes
    xs = []
    fs = []

    # the following is identical to our iteration above, apart from storing iterations
    x_prev = x + 2*atol
    while abs(x - x_prev) > atol:
        x_prev = x
        x = f(x_prev)
        print('x_{0:0>2d} = {1:7.5f}.   g(x_{0:0>2d}) = {2:7.5f}.   |x_{0:0>2d} - g(x_{0:0>2d})| = {3:7.5f}'.
              format(len(xs), x_prev, x, abs(x_prev-x)))
        xs.append(x_prev)
        fs.append(x)  # as x = g(x_prev) at this point we can store this

    # plot the iteration results
    ax.scatter(xs, fs, marker='x', color='red', s=30)

    # plot the convergence pattern
    x_pattern = [xs[0]]
    f_pattern = [fs[0]]
    for i in range(1, len(xs)):
        x_pattern.append(xs[i])
        f_pattern.append(fs[i-1])
        x_pattern.append(xs[i])
        f_pattern.append(fs[i])
    ax.plot(x_pattern, f_pattern, 'g--')

    # plot the function over the x values considered
    idx_sort = np.argsort(xs)
    ax.plot(np.array(xs)[idx_sort], np.array(fs)[idx_sort], 'b')

    # add some labels
    # figure out a reasonable offset for labels
    dxt = (np.max(x_pattern)-np.min(x_pattern))/50.
    dyt = (np.max(f_pattern)-np.min(f_pattern))/50.
    ax.text(xs[0]+dxt, fs[0]+dyt, '$x_0$')
    # only plot a maximum of max_labels labels, so plot doesn't get too messy
    for i in range(1, min(max_labels+1, len(xs))):
        label = ''.join(['$x_{', str(i), '}$'])
        ax.text(xs[i]+dxt, fs[i]+dyt, label)

    ax.set_xlabel('$x$', fontsize=16)
    if not flabel:
        fl = '$f(x)$'
    else:
        fl = flabel
    ax.set_ylabel(fl, fontsize=16)
    ax.set_title('Successive Approximations', fontsize=20)


def g(x):
    return np.exp(-x)

# set up a figure for plotting
fig, ax1 = plt.subplots(figsize=(8, 8))
# set an initial guess - you can try different values for x_guess
x_guess = 1.0
# call our plotting function with above choice of function g(.), use the default stopping criterion
plot_picard_convergence(g, x_guess, ax1, flabel=r'$g(x) = \exp(-x)$')

# add a y=x line without extending the axes limits
ax1.autoscale(False)
xlim = plt.gca().get_xlim()
ax1.plot(xlim, xlim, 'k--')
````

Recall that we introduced this so-called ["cobweb plot"](https://en.wikipedia.org/wiki/Cobweb_plot) in ACSE-2.

- We start from $x^{(0)}=1.0$ - shown as $x_0$ in the plot.


- We evaluate $g(x^{(0)}) = 0.36788...$. The red cross marked as $x_0$ is plotted at the location $(x^{(0)},g(x^{(0)}))$.


- We set our new iteration value, $x^{(1)}$, to this '$y$' value (i.e. $0.36788...$) - the lowest green dashed horizontal  line takes us from $x^{(0)}$ to this new $x^{(1)}$ - we jump horizontally to the $y=x$ line given by the black dashed line.


- We then evaluate $g(x^{(1)}) = 0.69220...$ - the left most vertical green dashed line takes us to this value on the $y$ axis where the second red cross marked with $x_1$ is plotted at the location $(x^{(1)},g(x^{(1)})$.


- We set our new $x$ value to this (0.69220), and this is indicated by the top most horizontal green dashed line taking us back to the $y=x$ black dashed line.


- Evaluate $g$ of this value giving us 0.50047, this is our next vertical line


- and so on, converging to our fixed point

This plot shows that our algorithm *converges* to the point where $x = g(x)$, which in our case is $x = \mathrm{e}^{-x}$.

For this example this is the point where both $x$ and $g(x)$ equal ~0.57. 

It should be obvious that this method is iterative in nature and that the solution can be expected to be an approximation accurate to a tolerance of $e \leq$ `atol`. 

As we said at the start of this lecture, another way to visualise this is to plot the functions $y = x$ and $y = \mathrm{e}^{-x}$ in $(x,y)$ space and find the intersection.

````python
fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)

x = np.linspace(0, 1, 100)

ax1.plot(x, np.exp(-x), 'b', label=r'$y(x) = g(x) = \mathrm{e}^{-x}$')
ax1.plot(x, x, 'k--', label='$y(x)=x$')
# our solution from Picard above: 0.5671430835570621
ax1.plot(0.5671430835570621, 0.5671430835570621, 'ro', label='intersection')
ax1.legend(loc='best', fontsize=14)
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel('$y(x)$', fontsize=16)
ax1.set_title('Fixed point as the intersection of two curves', fontsize=16)
````

## Roots of equations

This graphical method shows the intersection point, $x = \mathrm{e}^{-x}$, also at ~0.57. 

This value is also termed a *root* of a closely related problem, as it satisfies 

$$ 0 = f(x) := x - \mathrm{e}^{-x}. $$

[So a "fixed point" can always be defined as a "root" of a related problem, but the opposite is not strictly speaking true. However, we will see below how it is possible to update the problem of finding a "root" to finding a "fixed point", and hence reapply this sort of fixed point iteration approach to this more general problem].

In this expression we rearranged $\,x = \mathrm{e}^{-x}\,$ to $\,0 = x - \mathrm{e}^{-x}\,$ and defined $\,f(x) := x - \mathrm{e}^{-x}$. 

Thus, another solution strategy is to find a value $x^*$ such that $f(x^*) = 0$, or, in our case, $x^* - \mathrm{e}^{-x^*} = 0$. 

We can of course plot a similar intersection figure for this modified problem.

````python
fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)

x = np.linspace(-1, 2, 100)

ax1.plot(x, x-np.exp(-x), 'b', label='$y(x) = f(x) = x - \mathrm{e}^{-x}$')
# add a zero line extending across axes
xlim = ax1.get_xlim()
ax1.plot([xlim[0], xlim[1]], [0., 0.], 'k--', label='$y(x)=0$')
ax1.plot(0.5671, 0., 'ro', label='intersection')
ax1.set_xlim(xlim)
ax1.legend(loc='best', fontsize=14)
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel('$y(x)$', fontsize=16)
ax1.set_title('Fixed point as a root of $f(x)$', fontsize=16);
````

As expected the graph shows the root of $f(x)$ at our solution $x^* \approx 0.57$. 

Note also that by visual inspection, we see that $\,f(x)\,$ has a root $\,f(x^*) = 0\,$ in the interval $x^* \in (-1,1)$. This sort of observation will be important later.  

## Convergence analysis

Let's think a bit about whether or not we expect this method to converge.

Firstly let's plot the behaviour of the method for a problem that does indeed converge.

````python
def plot_picard_convergence2(f, x, ax, max_labels=6, atol=1.0e-4, max_its=4, flabel=''):
    """ Function to demonstrate/plot the Picard algorithm and its convergence.
    
    Updated slightly from plot_picard_convergence to impose a maximum number of iterations
    and plot the function.
    """
    # variables to store solution iterations for plotting purposes
    xs = []
    fs = []

    # the following is identical to our iteration above, apart from storing iterations
    x_prev = x + 2*atol
    it = 0
    while abs(x - x_prev) > atol:
        x_prev = x
        x = f(x_prev)
        xs.append(x_prev)
        fs.append(x)  # as x = g(x_prev) at this point we can store this
        it += 1
        if it==max_its:
            break

    # plot the iteration results
    ax.scatter(xs, fs, marker='x', color='red', s=30)

    # plot the convergence pattern
    x_pattern = [xs[0]]
    f_pattern = [xs[0]]
    for i in range(1, len(xs)):
        x_pattern.append(xs[i-1])
        f_pattern.append(fs[i-1])
        x_pattern.append(xs[i])
        f_pattern.append(fs[i-1])
    x_pattern.append(xs[i])
    f_pattern.append(fs[i])    
    ax.plot(x_pattern, f_pattern, 'g--')

    # plot the function over the x values considered
    xf = np.linspace(np.min(xs)-0.1, np.max(xs)+.1, 100)
    ax.plot(xf,f(xf),'b',label='$y = g(x)$')
    ax.plot(xf,xf,'k',label='$y = x$')
    ax.legend(loc='best', fontsize=14)

    # add some labels
    # figure out a reasonable offset for labels
    dxt = (np.max(x_pattern)-np.min(x_pattern))/50.
    dyt = (np.max(f_pattern)-np.min(f_pattern))/50.
    ax.text(xs[0]+dxt, fs[0]+dyt, '$x_0$')
    # only plot a maximum of max_labels labels, so plot doesn't get too messy
    for i in range(1, min(max_labels+1, len(xs))):
        label = ''.join(['$x_{', str(i), '}$'])
        ax.text(xs[i]+dxt, fs[i]+dyt, label)

    ax.set_xlabel('$x$', fontsize=16)
    if not flabel:
        fl = '$f(x)$'
    else:
        fl = flabel
    ax.set_ylabel(fl, fontsize=16)
    ax.set_title('Successive Approximations', fontsize=20)


def g(x):
    return np.sqrt(x)


fig, ax1 = plt.subplots(figsize=(8, 8))
x_guess = 0.4
plot_picard_convergence2(g, x_guess, ax1,  max_labels=3, atol=1.0e-2, max_its=6, flabel=r'$y$')
````

The example above shows clear convergence to the fixed point (this will always be the case for this particular problem as long as we start with a positive initial guess (think about why it needs to be positive, and check this claim by updating the initial guess), and more generally in situations which locally look like this as long as we start sufficiently close to the fixed point).

However, the following example shows a case where even if we start close to the fixed point, the algorithm actually diverges.

"Close" is of course a bit of a useless term here, since how do we judge if a guess is close or not - i.e. relative to what? 

So note that here that no matter how close to the fixed point we start, we move away from it! You could also check this with different initial guesses.

````python
def g(x):
    return np.tan(x) - .1


fig, ax1 = plt.subplots(figsize=(8, 8))
# start very close to root using value found below - need to set atol lower (i.e. strict) otherwise stops at first it.
x_guess = 0.63
plot_picard_convergence2(g, x_guess, ax1,  max_labels=12, atol=1.0e-6, max_its=12, flabel=r'$y$')
````

What's going on in these two cases? What's the difference that leads to different behaviour?

Basically it's the derivative of the function in the vicinity of the fixed point (the slope of the blue line).

If the magnitude of this is less than unity (slope less that that of the black line) then we will have convergence, but if it's larger than one we will have divergence.

Hopefully you can see that this to be the case from thinking about the problem geometrically as per the figures above (i.e. think about the progression of the green dashed line when the slope of the blue line is lower or higher than the slope of the black line, which is 1). 

But let's also arrive at the same conclusion from a mathematical analysis.

[Note that we will return to this failing example shortly, showing a way to recover convergence].

## Mathematical convergence analysis  [$\star$]

Our iteration is defined as

$$ x^{(k+1)} = g(x^{(k)}). $$

Let $x^*$ denote the fixed point of $g$, i.e.

$$ g(x^*) = x^*. $$

Denote the error at iteration $k$ as $e^{(k)}$ and define it mathematically as

$$e^{(k)} := x^* - x^{(k)} \quad\iff\quad  x^{(k)} = x^* - e^{(k)} , $$

so we also have 

$$e^{(k+1)}  = x^* - x^{(k+1)} \quad\iff\quad  x^{(k+1)} = x^* - e^{(k+1)} . $$

Substituting these two expressions into our iteration, $ x^{(k+1)} = g(x^{(k)}) $,  we have 

$$ x^* - e^{(k+1)} = g(x^* - e^{(k)}). $$

When we see something like the RHS [a function ($g$) evaluated at a point ($x^*$) plus a little bit ($-e^{(k)}$)], an immediate thought should be to consider what a Taylor series expansion tells us. 

As we know something about $g(x^*)$, let's use Taylor series to expand the function $g$ about $x^*$:


$$g(x) = g(x^*) + g'(x^*) (x - x^*) + \frac{g''(x^*)}{2!} (x - x^*)^2 + \mathcal{O}((x - x^*)^3),$$

and hence the RHS of the above expression involving errors ($ x^* - e^{(k+1)} = g(x^* - e^{(k)}) $) can be written as

$$g(x^* - e^{(k)}) = g(x^*) - g'(x^*) e^{(k)} + \frac{g''(x^*) (e^{(k)})^2}{2!} + \mathcal{O}((e^{(k)})^3),$$

with the full expression therefore being equivalent to

$$x^* - e^{(k+1)} = g(x^*) - g'(x^*) e^{(k)} + \frac{g''(x^*) (e^{(k)})^2}{2} + \mathcal{O}((e^{(k)})^3).$$


Since $x^* = g(x^*)$ we have the following expression for the evolution of the error with iteration

$$e^{(k+1)} = g'(x^*) e^{(k)} + \mathcal{O}((e^{(k)})^2).$$

In an [*asymptotic analysis*](https://en.wikipedia.org/wiki/Asymptotic_analysis) we assume that we are in some limiting behaviour. Here we assume we are close enough to the root such that two things are true.

The first is that $e^{(k)}$ is small enough such that we can ignore the second order (in $e^{(k)}$) terms, i.e. we can assume that 

$$e^{(k+1)} \approx g'(x^*) e^{(k)}.$$


The second is that provided that $g'(x^*)$ is bounded in magnitude below unity, then there exists a constant $C<1$ such that [NB. we can do this a bit more correctly if we make use of a Taylor remainder expression]

$$ |e^{(k+1)}| \leq C |e^{(k)}| \implies |e^{(k)}| \leq C^k |e^{(0)}|,$$

which, as $C<1$, tends to zero with $k$, i.e. we have convergence.

Moreover this analysis shows us that the convergence is *linear* (it drops by the same factor every iteration), and at a rate proportional to the magnitude of $g'(x^*)$. Smaller magnitude derivatives will have faster convergence.

This can be called a local convergence analysis as we have assumed that we are *local* to the fixed point.

More generally we will often find error relationships of the form

$$|e^{(k+1)}| \leq C |e^{(k)}|^p.$$

In the $p=1$ case, if $C\geq 1$ then clearly a method with this property will not converge (assuming that $C$ represents a tight bound on the magnitude of the derivative at the root); furthermore if $C > 1$ then the error will actually grow every iteration and the method will diverge.

If $C < 1$ then the method will converge as the error will then decay every iteration.

Above we saw an example where $p=1$, in this case the error will decrease by the factor $C$ every iteration, i.e. convergence is linear. E.g. if $C = 1/2$ the error will drop by a factor of two every iteration. Note therefore that if $C$ is only slightly smaller than unity then convergence will be slow and we may need to use a lot of iterations to reach our convergence tolerance.

Below we will see examples where $p=2$, which we refer to as converging quadratically. Here for convergence we require that the error is small enough to start with, not that $C$ is less than one.

Let's plot how errors that obey this relationship drop with iteration number to see why we are so interested in quadratically converging schemes!

````python
def convergence_example(C, n_its):
    e1 = np.zeros(n_its)
    e2 = np.zeros(n_its)
    e1[0] = 0.1
    e2[0] = 0.1
    for it in range(n_its-1):
        e1[it+1] = C * e1[it]**1
        e2[it+1] = C * e2[it]**2
    return e1, e2
    
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
fig.tight_layout(w_pad=4, h_pad=4)

for i, C in enumerate((0.9, 0.5, 0.01)):
    e1, e2 = convergence_example(C, 10)
    ax[i].semilogy(e1,'k.-',label='linear')
    ax[i].semilogy(e2,'b.-',label='quadratic')
    ax[i].set_ylim([1e-16,1])
    ax[i].set_xlabel('iteration number', fontsize=14)
    ax[i].set_ylabel('log(error)', fontsize=14)
    ax[i].set_title('Convergence ($C={}$)'.format(C), fontsize=16)
    ax[i].legend(loc='best', fontsize=14)
````

Of course there's rarely a "free lunch" and the problem with quadratically converging schemes is that they generally require a very good starting guess.

A robust approach with good performance is therefore to combine methods - e.g. use a linear method to get close to the root, and then a quadratic method will converge to a very tight tolerance in perhaps only a few further iterations.

# Subintervals

## An example

Let's consider an example:

$$2\,x + x \, \sin(x-3) = 5\;\;\;\; \text{for}\;\;\;\; x \in (-10,10).$$


By means of visual inspection let's find a subinterval $(a,b)$ such that 


1. there exists an $x^* \in (a,b)$ such that $f(x^*) = 0$, and 


2. $f(x)$ is [monotonic](https://en.wikipedia.org/wiki/Monotonic_function).

Where we define $f$ such that the solution to the above equation is a root - i.e. move all the terms on to one side and set equal to zero. Below we make the choice to define $f$ as $f(x) = 2\,x + x \, \sin(x-3) - 5$.

````python
# for this problem we can define f as 5 - (2*x + x*np.sin(x-3)), or (2*x + x*np.sin(x-3)) - 5.
# it makes no difference to the root-finding, just the plots
def f(x):
    return 2*x + x*np.sin(x-3) - 5


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.tight_layout(w_pad=4, h_pad=4)

x = np.linspace(-10. ,10. , 100)
ax1.plot(x, f(x),  'b', label='$y(x) = f(x)$')
xlim = ax1.get_xlim()
ax1.plot([xlim[0], xlim[1]], [0., 0.], 'k--', label='$y(x)=0$')
ax1.plot(2.7903546180675676, 0., 'ro', label='intersection')
ax1.set_xlim(xlim)
ax1.legend(loc='best', fontsize=14)
ax1.set_xlabel('$x$', fontsize=14)
ax1.set_ylabel('$y$', fontsize=14)
ax1.set_title('Fixed point as a root of $f(x)$ - wider $x$ range', fontsize=14)

x = np.linspace(0., 5., 100)
ax2.plot(x, f(x),  'b', label='$y(x) = f(x)$')
xlim = ax2.get_xlim()
ax2.plot([xlim[0], xlim[1]], [0., 0.], 'k--', label='$y(x)=0$')
ax2.plot(2.7903546180675676, 0., 'ro', label='intersection')
ax2.set_xlim(xlim)
ax2.legend(loc='best', fontsize=14)
ax2.set_xlabel('$x$', fontsize=14)
ax2.set_ylabel('$y$', fontsize=14)
ax2.set_title('Fixed point as a root of $f(x)$ - narrower $x$ range', fontsize=14);
````

It seems that in $(0,5)$, i.e. an interval with upper and lower bounds $a$ = 0 and $b$ = 5, there exists a root, and also the function is monotonically increasing over this interval - we have excluded any local maxima or minima that we see do occur in the function outside this restricted interval.

[Note the caveat that in general we will assume the evaluation of $\,f$ (or $g$) is expensive and so we want to do this as few times as possible - generating a nice high resolution plot like this is not something we will be able to do in general].

We can make use of this knowledge to help us identify a good starting guess for the root.

## An application of Picard iteration close to the root

Let's see if we can use Picard iteration to find the root.

But first a note of caution, we know from above that in terms of the functon $g$ we require that it's derivative at the fixed point isn't too large to ensure convergence.

the function $f$ we have here is a potential issue since it's derivative looks steep near the root.

Let's actually compute the value of the derivative at the root:

````python
def f(x):
    return 2*x + x*np.sin(x-3) - 5

# use a SciPy nonlinear solver, with initial guess of 3 based on figures above
x_sop = sop.newton(f, 3.0)

print(x_sop)

# compute the derivative by hand
def dfdx(x):
    return 2 + np.sin(x-3) + x*np.cos(x-3)

# and evaluate it at the root
print(dfdx(x_sop))
````

This value is quite large and so we may have problems.

Let's check our intuition/understanding here is correct.  Of course to call our Picard iteration code, we need to think about how to define the function $g$.

Notice that

$$f(x) = 0 \iff x = x + f(x), $$

and so one options if to define $g(x) := x + f(x)$.  Alternatively

$$f(x) = 0 \iff x = x - f(x), $$

and so $g(x) := x - f(x)$ is another viable option.

Adding AND subtracting an $x$ may seem a bit of a weird thing to do, but this is simply so we get an $x$ on its own placing the problem in the form that our fixed point iteration algorithms/code requires. Let's apply our Picard iteration to these two $g$ functions.

````python
def f(x):
    return 2*x + x*np.sin(x-3) - 5

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.tight_layout(w_pad=4)

x_guess = 2.7


# as Picard solves 'x = g(x)', reformulate the problem 'f(x) = 0' to 'x = g(x) := x - f(x)'' 
def g(x):
    return x - f(x)

plot_picard_convergence2(g, x_guess, ax1,  max_labels=3, atol=1.0e-2, max_its=3, flabel=r'$y$')

# of course it shouldn't make any difference if we define g(x) := x + f(x) - let's confirm in a plot
def g(x):
    return x + f(x)

plot_picard_convergence2(g, x_guess, ax2,  max_labels=3, atol=1.0e-2, max_its=3, flabel=r'$y$')
# ... the details are of course different, but the story is the same.
````

So even though we did start quite close to the root (or fixed point of the problem rewritten in terms of $g$ - and note that two possibilities for how to do this were tested in the two frames above), our Picard iteration indeed diverges as we suspected it would.

But this flexibility over how we formulate the problem (i.e. which function $g$ we pass to our Picard iteration code) opens up a possibility where we might be able to change the magnitude of the gradient of $g$, which we know from above dictates the convergence vs divergence of our iteration (and assuming convergence, also the convergence rate).

It doesn't make any difference to our problem specification if instead of adding and subtracting $x$ from $f$, we add and subtract a multiple of $x$, or equivalently add and subtract $x$ to a multiple of $f$ - let's see if it makes a difference to the algorithms convergence:

````python
def f(x):
    return 2*x + x*np.sin(x-3) - 5

# of course we are also free to multiply the f(x) here by a constant, as at convergence
# we want it to be zero - but f(x) being zero is of course equivalent to m*f(x) = 0, so
# 'g(x) := x - m*f(x)' is another possible option.
# Let's see what happens with a few choices of a multiplier m.

# define three g functions with different choice of the multiplier m:
def g1(x):
    return x - 0.6*f(x)

def g2(x):
    return x - 0.4*f(x)

def g3(x):
    return x - 0.2*f(x)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.tight_layout(w_pad=4)
x_guess = 2.7

plot_picard_convergence2(g1, x_guess, ax1,  max_labels=3, atol=1.0e-2, max_its=3, flabel=r'$y$')
plot_picard_convergence2(g2, x_guess, ax2,  max_labels=5, atol=1.0e-2, max_its=5, flabel=r'$y$')
plot_picard_convergence2(g3, x_guess, ax3,  max_labels=3, atol=1.0e-2, max_its=3, flabel=r'$y$')
ax1.set_title('Divergence', fontsize = 20)
ax2.set_title('Slow convergence', fontsize = 20)
ax3.set_title('Fast convergence!', fontsize = 20);
````

So choice of the multiplier $\,m\,$ takes us from divergence, through slow convergence to very rapid convergence.

Choice of a good multiplier can therefore be of critical importance. Let's develop this idea in the next section.

# The chord method

So it appears that we can recover convergence through suitable definition of the function $g$, i.e. an appropriate scaling of $f$ in the definition of the $g$ we give to our Picard iteration code.


This choice can clearly also have a dramatic impact on the convergence rate (compare the 2nd and 3rd frames above)!


This method - the application of Picard (fixed point) iteration to the problem

$$ x = g(x):= x-mf(x) \;\;\;\; \text{i.e. the iteration} \;\;\;\; x^{(k+1)} = g(x^{(k)}) = x^{(k)} - mf(x^{(k)}), $$

is sometimes called the *chord method*. 

We established above that for our iteration to converge we need $|g'(x)| < 1$ around our fixed point. 

With $\,g(x) := x - mf(x)\,$ for some $\,m\,$, this criterion can be rewritten as 


$$| 1 - m f'(x)| < 1 \iff -1 < 1 - m f'(x) < 1 \iff 0 <  m f'(x) < 2. $$


This criterion gives us a mechanism to choose an appropriate $m$ (and explains the behaviour we observed for the example above given we found the derivative of $f$ at the fixed point to be approx 4.5).

Note that we are also free to choose different $m$ at each iteration:

$$x^{(k+1)} = x^{(k)} - m_k f(x^{(k)}). $$

We will see examples of this below. 

An optimal choice of an iteration-dependent $m$ actually leads to a method with quadratic convergence. We shall also see that below.

But first let's check if this theory allows us to save our failing problem from above.

Recall that the problem were we first plotted divergent behaviour was where we were trying to find the fixed point (i.e. $x$ such that $x=g(x)$) where

$$ g(x) := \tan(x) - \frac{1}{10}.$$

````python
def g(x):
    return np.tan(x) - .1

fig, ax1 = plt.subplots(figsize=(8, 8))
# start very close to root using value found below - need to set atol lower (i.e. strict) otherwise stops at first it.
x_guess = 0.63
plot_picard_convergence2(g, x_guess, ax1,  max_labels=12, atol=1.0e-6, max_its=12, flabel=r'$y$')
````

As above we can re-pose the problem as finding a root, i.e. $\,f(x) = 0\,$, where for this example

$$ f(x) := x - g(x) = x - \left(\tan(x) - \frac{1}{10}\right), $$

the chord method then iterates $\, x^{(k+1)} = x^{(k)} - mf(x^{(k)})\, $ and for convergence we want to choose $m$ such that

$$0 <  m f'(x) < 2. $$

For this example 

$$f'(x) = 1 - \frac{d}{dx}\left(\tan(x)\right) = 1 - \frac{1}{\cos^2(x)},$$

which we need to evaluate near our fixed point - we know from the (failing Picard) plot above that the fixed point is located at somewhere between $x=0.6$ and $x=0.7$.

Let's evaluate $f'(x)$ at the two locations to give an idea of the bounds on the potential $m$ we can use.

````python
a = 0.6
b = 0.7
print( 1. - (np.cos(a)**(-2.) ))
print( 1. - (np.cos(b)**(-2.) ))


# of course we can cheat and actually compute the root and the value of f' at the root:
def f(x):
    return x - (np.tan(x) - .1)

x_sop = sop.newton(f, 0.65)
print('\nSolution from SciPy: ',x_sop)
print("f'(root) = ", 1. - (np.cos(x_sop)**(-2.) ))
print("2./ f'(root) = ", 2./(1. - (np.cos(x_sop)**(-2.) )))
````

So

$$1 - \frac{1}{\cos^2(a)} = -0.46804317252795724, $$

and 

$$1 - \frac{1}{\cos^2(b)} = -0.709449715863117. $$

These are both negative and the largest magnitude value is ~-0.71.

For convergence we require that 

$$0 <  m f'(x) < 2, $$

which for this problem we can replace with

$$0 <  -0.71 m  < 2 \iff 0 > m > \frac{2}{-0.71} \approx -2.8. $$

By choosing an upper estimate on the derivative this gives us a conservative estimate for $m$.

````python
def g(x):
    return np.tan(x) - .1

def f(x):
    return x - g(x)

# this should recreate the lack of convergence plot we saw earlier
def g1(x):
    return x - 1.0*f(x)

# this defn of g should converge based on our application of our theory above (m=-2.8; 
# values between 0 and this should also converge - you could check this)
def g2(x):
    return x + 2.8*f(x)

# and this may not (depends on how representative our estimate of the 
# derivative at the fixed point was)
def g3(x):
    return x + 3.5*f(x)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.tight_layout(w_pad=4)
x_guess = 0.6

plot_picard_convergence2(g1, x_guess, ax1,  max_labels=5, atol=1.0e-2, max_its=5, flabel=r'$y$')
plot_picard_convergence2(g2, x_guess, ax2,  max_labels=3, atol=1.0e-2, max_its=5, flabel=r'$y$')
plot_picard_convergence2(g3, x_guess, ax3,  max_labels=5, atol=1.0e-2, max_its=5, flabel=r'$y$')
````

To get a better understanding of what's going on, let's plot the convergence behaviour of the method in terms of iterations towards the root of the problem $f(x)=0$ with different choices of $m$.

````python
def plot_chord(f, m, x_0, ax, max_labels=6, atol = 1.e-5, maxiter=100, resfct = 100, flabel=''):
    x_n = [x_0]
    y_n = [f(x_0)]    
    
    # Chord iteration
    i = 0
    # iterate until we hit break either as we hit tolerance or maximum number iterations
    while 1:
        # Chord update
        x_zero = x_n[-1] - m * f(x_n[-1])
        x_n.append(x_zero)        
        y_n.append(0.)
        if abs(x_n[-1]-x_n[-2]) < atol:
            break
        x_n.append(x_zero)
        y_n.append(f(x_zero))
        i = i+1
        if i >= maxiter:
            break
        
    # the iteration results
    ax.scatter(x_n, f(np.array(x_n)), marker='x', color='red', s=30)
    
    # the convergence pattern
    ax.plot(x_n, y_n, color='green', ls='--')

    # add some labels
    # figure out a reasonable offset for labels
    dxt = (np.max(x_n)-np.min(x_n))/50.
    dyt = (np.max(f(np.array(x_n)))-np.min(f(np.array(x_n))))/50.
    ax.text(x_n[0]+dxt, f(x_n[0])+dyt, '$x_0$')
    # only plot a maximum of max_labels labels, so plot doesn't get too messy
    for i in range(1, min(max_labels+1, len(x_n)), 2):
        label = ''.join(['$x_{', str(np.int((i+1)/2)), '}$'])
        ax.text(x_n[i]+dxt, f(x_n[i])+dyt, label)
        
    # the function
    x = np.linspace( np.min(x_n) - 0.01, np.max(x_n) + 0.01, resfct)
    ax.plot(x, f(x), 'b')
    
    # zero line
    xlim = ax.get_xlim()
    ax.plot([xlim[0], xlim[1]], [0., 0.], 'k--')
    ax.set_xlim(xlim)    
    ax.set_xlabel('$x$', fontsize=16)
    if not flabel:
        fl = '$f(x)$'
    else:
        fl = flabel
    ax.set_ylabel(fl, fontsize=16)
    ax.set_title("Chord method iteration", fontsize=20)
    # add a triangle to indicate the line slope (with the slope being 1/m!)
    xc = (np.max(x_n) + np.min(x_n))/2.
    yc = (np.max(f(np.array(x_n))) + np.min(f(np.array(x_n))))/2.
    annotation.slope_marker((xc, 2*yc), (1./m, 1.), invert=False, ax=ax,
                        size_frac=0.15, pad_frac=0.05)


def g(x):
    return np.tan(x) - .1

def f(x):
    return x - g(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
x0 = 0.6
m = -2.8
plot_chord(f, m, x0, ax1, atol=1.e-2, maxiter=4, flabel=r'$y$')
print('The slope marker contains the value 1/m = ',1./m)

m = -1.
plot_chord(f, m, x0, ax2, atol=1.e-3, maxiter=5, flabel=r'$y$')
print('The slope marker contains the value 1/m = ',1./m)
````

Our iteration here is

$$x^{(k+1)} = x^{(k)} - m f\left(x^{(k)}\right), $$

which we can rearrange to 

$$ \frac{1}{m} = \frac{0 - f\left(x^{(k)}\right)}{x^{(k+1)} - x^{(k)}}.$$

This explains why the slopes of the updates (the diagonal green dashed lines) in the plots above are $1/m$.

We will come back to this equation and these ideas when we introduce Newton's method below.

# Root bracketing

We've seen several times now that having an approximate idea of where the root lies can be important.

By visual inspection we can identify if roots exist for a function, and narrow down the interval in which a root can be found. 

For the equation 

$$f(x) := x - \text{e}^{-x},$$ 

we have shown by plotting (above in two images, see also below, where we noted a root around $x\approx ~0.57$) that a root is bounded in $(-1,1)$, and that $f(x)$ is monotonically increasing over this interval. 

With this in mind, we can define a [*root bracketing algorithm*](https://en.wikipedia.org/wiki/Root-finding_algorithm#Bracketing_methods) that marches along $f(x)$ in increments of $\Delta x$ and identifies a new, tighter (i.e. smaller) bracket around the root by detecting a *change in sign of $f(x)$*. 

This algorithm is also referred to as *incremental search*. 

The approach can be visualised as follows:

````python
def plot_root_bracketing(f, a, b, dx, ax, xbounds=(-0.1, 1.4), ybounds=(-5, 6), flabel=''):
    x = np.linspace(a, b, int((b-a)/dx)+1)
    y = f(x)
    # plot the sub-intervals in blue
    ax.plot(x, y, 'bo-')
    for i in range(1, len(x)):
        if np.sign(y[i]) != np.sign(y[i-1]):
            # plot the sub-interval where the sign changes in red
            ax.plot([x[i], x[i-1]], [y[i], y[i-1]], 'ro-')
    ax.set_xlabel('$x$', fontsize=16)
    if not flabel:
        fl = '$f(x)$'
    else:
        fl = flabel
    ax.set_ylabel(fl, fontsize=16)
    xlim = ax.get_xlim()
    ax.plot([xlim[0], xlim[1]], [0., 0.], 'k--')
    ax.set_xlim(xlim)
    ax.set_title('Root bracketing\n' + '(red indicates the bracket containing the root)', fontsize=20)

    
def f(x):
    return x - np.exp(-x)


fig, ax1 = plt.subplots(figsize=(8,5))
    
lower_bound = -1.
upper_bound = 1.
dx = 0.4
plot_root_bracketing(f, lower_bound, upper_bound, dx, ax1, flabel=r'$f(x) = x - \mathrm{e}^{-x}$')
````

## Root bracketing example
    
For our example from above

$$2\,x + x\, \sin(x-3) = 5,$$

let's use the subinterval $x \in(a,b)$ we found above  to implement a root bracketing algorithm. 

````python
def f(x):
    return 2*x + x*np.sin(x-3) - 5


def root_bracketing(f, a, b, dx):
    """ Function to perform root bracketing on the function f(.)
    between a and b, with fixed interval size dx.
    Returns the bracket of size dx that contains the root.
    """ 
    # The sign function returns:  -1 if x < 0;  0 if x==0;  1 if x > 0.
    sign = np.sign(f(a))
    while sign == np.sign(f(a)):
        a += dx
        if a >= b:
            raise RuntimeError('no root within [a,b]')
    return (a-dx, a)


a = 0.
b = 5.
dx = 0.1
# print out the output from our root_bracketing function
print('Bracket = ', root_bracketing(f, a, b, dx))

# let's also use our plotting function from above.
fig, ax1 = plt.subplots(figsize=(8,5))
plot_root_bracketing(f, a, b, 0.1, ax1, flabel=r'$f(x) = 2x + xsin(x-3) - 5$')
````

So we now have an even tighter handle on where the root is. 

We can use this knowledge when giving an initial guess to later methods.

Note that as the size of the intervals ($\Delta x$) was fixed, this method is not iterative and hence does not converge in any real sense to the root.

However, we can extend this idea by developing a method which incrementally refines the sub-interval size - as we shall now see.

# Bisection method

Once we know that a root can be found in $x \in (a,b)$, we can close in on it with an algorithm similar to root-bracketing/incremental search described above, but with a smart switch to gradually *decrease* $\Delta x$ *by a factor* of 1/2, and *change the marching direction* depending on the sign of $f(x_1)f(x_2)$, where $x_1$ and $x_2$  are the local bounds considered during the marching process.


The algorithm works as follows:


- If there is a root in the interval $[x_1, x_2]$, then $f(x_1)f(x_2) < 0$ (as $f(x_1)$ and $f(x_2)$ will be of different signs)


- In order to halve the interval, we compute $f(x_3)$, where $x_3 := (x_1 + x_2)/2$ is the midpoint of the current interval. 


- If $f(x_2)f(x_3) < 0$, then the root must be in $[x_2, x_3]$, and we record this by replacing the original bound $x_1$ by $x_3$,


- otherwise, the root must lie in $[x_1, x_3]$, in which case $x_2$ is replaced by $x_3$. 


- In either case, the new updated interval $[x_1, x_2]$ is half the size of the original interval. 


- The bisection is repeated until the interval has been reduced to some user-defined convergence tolerance.

## Convergence analysis

Note that since the size of the bracket only halves at each iteration (i.e. $\Delta x_{k+1} = \Delta x_{k}/2$),  we can only know the root to within an error of magnitude $\Delta x_k$ at each iteration.

Hence we have

$$ |e_{k}| = \Delta x_{k}\;\;\;\;\;\text{and}\;\;\;\;\;|e_{k+1}| = \Delta x_{k+1} \;\;\;\;\;\text{where}\;\;\;\;\; \Delta x_{k+1} = \frac{1}{2} \Delta x_{k},$$

which together implies that 

$$|e_{k+1}| = \frac{1}{2} |e_k|,$$

i.e. the method converges linearly.

# Newton's method

So far, the above iterative algorithms use only one part of a function's information: its value, $\,f(x)$. 

*Newton's method* additionally uses the function's derivative $f'(x)$, using the slope of the function in the vicinity of $x$ to get closer to the root more optimally at every iteration. 

It's this additional information that yields a quadratically converging scheme. We saw and noted above just how valuable quadratic convergence can be.

The function derivative (slope $f'(x)$), together with the function value $f(x)$, are used to find the intersection of the tangent of $f$ at $x$ with zero (the $x$-axis) to obtain an improved guess of the root - note the similarity here with the chord method we saw above.

## Algorithm derivation (based on extending the chord method)

Given our introduction, analysis and (visual) investigation of the chord method above we can essentially write down Newton's method now, with a good-iteration dependent choice for the scaling parameter $m$ leading to very fast convergence.


Recall that under the chord method we could rearrange our iteration to

$$ \frac{1}{m} = \frac{0 - f(x^{(k)})}{x^{(k+1)} - x^{(k)}},$$

where we defined $m$ in advance (and it was fixed). 

Now we will use this interpretation of the iteration to come up with a good (non-fixed) $m$.

The zero we included explicitly in the expression above accounts for the fact that visually the chord update took us to zero (i.e. the $x$-axis) in one iteration.

Next we repeat the chord method plots we showed above:

````python
def g(x):
    return np.tan(x) - .1

def f(x):
    return x - g(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
x0 = 0.6
m = -2.8
plot_chord(f, m, x0, ax1, atol=1.e-2, maxiter=4, flabel=r'$y$')
print('The slope marker contains the value 1/m = ',1./m)

m = -1.
plot_chord(f, m, x0, ax2, atol=1.e-3, maxiter=5, flabel=r'$y$')
print('The slope marker contains the value 1/m = ',1./m)
````

Assuming our iteration converges and we are close to the root, $\,f(x^{(k+1)})\,$ (the value on the blue line in above figures) should be closer to zero than $\,f(x^{(k)})\,$. Hence we can approximate the above line with 

$$ \frac{1}{m} \approx \frac{f(x^{(k+1)}) - f(x^{(k)})}{x^{(k+1)} - x^{(k)}},$$

which again assuming we are near convergence, and so $x^{(k+1)} - x^{(k)}$ is small (and thus the RHS of the above approximates the derivative of $f$), tells us that a sensible choice for $m$ could well be

$$ \frac{1}{m} := f'(x^{(k)}). $$

Our chord update then reads 

$$x^{(k+1)} = x^{(k)} - \frac{ f(x^{(k)})) }{f'(x^{(k)})} , $$

and the method does what the chord method does in its iterations, but the direction of the green updated will be in the direction of the slope of the blue lines. Hopefully you can see for the two examples above this should lead to faster convergence.

## Algorithm derivation (based on Taylor series)

An alternative derivation (and the one probably most common in textbooks) is as follows.

An iteration formula for Newton's method can be derived by starting from the Taylor series expansion about the point $x^{(k)}$:

$$ f(x^{(k+1)}) = f(x^{(k)}) + f'(x^{(k)})(x^{(k+1)}-x^{(k)}) + \mathcal{O}\left(\left(x^{(k+1)} - x^{(k)}\right)^2\right), $$

i.e. we know that the following holds approximately (as long as $x^{(k+1)}$ is close to $x^{(k)}$):

$$f(x^{(k+1)}) \approx f(x^{(k)}) + f'(x^{(k)})(x^{(k+1)}-x^{(k)}).$$

Given an $x^{(k)}$, we would like an updated approximation to the root $x^{(k+1)}$ (i.e. we want $f(x^{(k+1)})=0$ - this is our aspiration, we won't necessarily achieve this in one step of course!).

So let's set this desired property of $x^{(k+1)}$ in the above relation to yield

$$ 0 = f(x^{(k)}) + f'(x^{(k)})(x^{(k+1)}-x^{(k)}). $$

Rearranging yields the update formula

$$ x^{(k+1)} = x^{(k)} - \frac{f(x^{(k)})}{f'(x^{(k)})}, $$

which is generally known as the **Newton-Raphson formula**.

And which to reiterate is in exactly the format of the chord method with the choice $m=1/f'(x)$.

A pseudo-code for the resulting iterative algorithm looks like this:

```
guess x[0]
while ( abs(x[k] - x[k-1]) > tolerance ) do
    x[k] = x[k-1] - f(x[k-1])/dfdx(x[k-1])
```

The following expression for the error relationship between iterations can be derived in a similar manner to our analysis of the Picard iteration by manipulating the Taylor series analysis:

$$ e^{(k+1)} \approx -\frac{f''(x^{*})}{2f'(x^{*})} \left(e^{(k)}\right)^2, $$

where $e^{(k)} := x^{*} - x^{(k)}$, with $x^{*}$ the root.

[Compare this with the result of our analysis of the convergence of Picard's method above].


### Comments

We can note a few things from this:


1. The smaller the second-derivative of $f$ compared to the first derivative (across the region we are iterating over) the smaller our errors will decay with iteration number (as this defines our convergence parameter $C$ we discussed above).


2. Assuming that $f$ is sufficiently well-behaved this demonstrates that the algorithm *converges quadratically*. Or said another way, the more complex (i.e. less smooth) the function the closer our initial guess will need to be to the root. Imagine a linear function (or a region containing the root where the function is linear), then based both on what the algorithm is doing as well as this error analysis, we would expect Newton to converge to the exact root in a single iteration!


3. But as we will see below, things go wrong if we don't have a sufficiently good guess.

## An alternative derivation [$\star$]

Note that looking at the convergence pattern plots above (and again recalling the chord method) motivates a slightly different approach to deriving the Newton-Raphson iteration update.

Starting from a current iteration location of $\,x^{(k)}$, and assuming that we have knowledge of both $\,f(x^{(k)})\,$ and $\,f'(x^{(k)})\,$, we can form an equation for the line which passes through the point $\,(x^{(k)},\, f(x^{(k)}))$, and which has slope $\,f'(x^{(k)})$.

The general equation for a line with slope $\,f'(x^{(k)})\,$ is:

$$ y(x) = f'(x^{(k)}) x + c, $$

where $c$ is the intercept. We can choose (or solve for) the intercept so the line passes through $\,(x^{(k)},\, f(x^{(k)}))$:

$$\begin{align*}
f(x^{(k)}) = y & = f'(x^{(k)}) x_i + c\\[5pt]
\implies c & = f(x^{(k)}) - f'(x^{(k)}) x^{(k)},
\end{align*}$$

Therefore the equation for our line (the green dashed diagonal lines in the convergence plots above) is

$$y(x) = f'(x^{(k)}) x + (f(x^{(k)}) - f'(x^{(k)}) x^{(k)}) = f'(x^{(k)}) (x - x^{(k)}) + f(x^{(k)}).$$

Our updated iteration value ($x^{(k+1)}$) for the estimated root is given by the location where this line intersects the $x$-axis (i.e. where $y = 0$):

$$\begin{align*}
0 = y(x^{(k+1)}) & = f'(x^{(k)}) (x^{(k+1)}-x^{(k)}) + f(x^{(k)})\\[5pt]
\implies x^{(k+1)} & = x^{(k)} - \frac{f(x^{(k)})}{f'(x^{(k)})},
\end{align*}$$

and we are left with exactly the **Newton-Raphson formula** we derived from Taylor series above.

If you're confused as you think we're simply saying the same thing in three different ways - don't worry that's exactly what we're doing. The aim is to re-enforce the method, and point our that there are almost always different approaches to think about or derive computational algorithms.

## Implementation

and of course we have already used it above in a couple of places without explanation!

````python
def f(x):
    return x - np.exp(-x)


def dfdx(x):
    return 1 + np.exp(-x)


x0 = -1.  # initial guess
# print the output of newton - we give it the function, an initial guess and the function derivative
print(sop.newton(f, x0, dfdx))

# note that this function also works if we don't give if dfdx (we will see below what it does in this situation)
print(sop.newton(f, x0))
````

Let's implement Newton's method ourselves in a function and compare with the SciPy version.

Note we haven't bothered entering a maximum number of allowed iterations here or in later functions, but this of course would be good practice and could follow the template used for bisection above.

````python
def newton(f, x0, dfdx, atol=1.0e-6):
    """ Function to implement the Newton-Raphson method
    
    f is the function we are trying to find a root of
    
    and dfdx is another function which return the derivative of f
    """
    x = [x0]
    fevals = 0
    k = 0
    while True:
        x.append(x[-1] - f(x[-1])/dfdx(x[-1]))  # two function evaluations (f and dfdx)
        fevals += 2
        k+=1
        if abs(x[-1]-x[-2]) < atol:
            print('Newton (analytical derivative) used', fevals, 'function evaluations')
            return x[-1],k
        
        
###### case 1
def f(x):
    return x - np.exp(-x)

def dfdx(x):
    return 1 + np.exp(-x)

x0 = -1.  # initial guess
print(newton(f, x0, dfdx))
print(sop.newton(f, x0, dfdx))


###### case 2
def f(x):
    return 2*x + x*np.sin(x-3) - 5

def dfdx(x):
    return 2 - np.sin(3-x) + x*np.cos(3-x)

x0 = 0.  # initial guess
print(newton(f, x0, dfdx))
print(sop.newton(f, x0, dfdx))
````

## Basins of attraction

A basin of attraction, or region of quadratic convergence, is the region within which our initial guess needs to be for our algorithm to converge to our desired root.

We have a section below giving examples of how we can fail to converge.

# But note that we have introduced slower converging method, e.g. bisection, that can get us close to the root or equivalently ensure we have a good guess and thus are inside the correct basin of attraction. 

## (Quasi-) Newton with approximate derivative 
    
The implementation of the Newton method above assumes that the derivative $f^\prime(x)$ is readily available. 

[Although note that this was not the case in the plotting function `plot_newton` where we computed an approximation to it].

For many problems, however, the derivative is not easy to express analytically or we just don't want to bother working it out and writing a function to implement it. 

In these cases $f^\prime(x)$ can be replaced by a difference approximation such as

$$ f'(x) \approx \frac{f(x+\Delta x) - f(x)}{\Delta x}. $$

Use of an approximate derivative renders the corresponding root finding algorithm a co-called *quasi-Newton* method.


Note that since we only have an approximate derivative we cannot in general expect this method to converge as well as Newton when we supply it with an exact derivative. This manifests in a reduction from perfect quadratic convergence - but still generally closer to quadratic that linear!


See the homework exercise for an implementation.


# Secant method [$\star$]

The [*Secant method*](https://en.wikipedia.org/wiki/Secant_method) is an alternative, closely-related approach which replaces the local derivative in the Newton method by a difference approximation based on two consecutive $x_n$ iteration values, i.e. makes use of function evaluations we have already undertaken, rather than having to compute additional evaluations. 

That is we approximate the derivative via

$$ f'(x^{(k)}) \approx \frac{f(x^{(k)}) - f(x^{(k-1)})}{x^{(k)}- x^{(k-1)}}, $$

which leads to the so-called *secant method* (remember we divide through by the derivative, so one over the expression above appears):

$$ x^{(k+1)} = x^{(k)} - f(x^{(k)}) \left ( \frac{x^{(k)} - x^{(k-1)}}{f(x^{(k)}) - f(x^{(k-1)})} \right ). $$

An advantage here is that we do not need to specify a sensible $\Delta x$ value with which to evaluate the numerical derivative.

An error analysis can be used to obtain

$$ e^{(k+1)} \approx C \, \left(e^{(k)}\right)^{1.618\ldots}, $$

where $1.618\ldots$ is the *golden ratio*, i.e convergence is far better than linear (hence it is what is called *super-linear*), but not quite as good as a fully quadratic method.

Aside: Note that since this method makes use of two previous iteration values to find the next, it does not fit within the framework we constructed around the chord method above. 

Also, we need to provide two values ($x^{(0)}$ and $x^{(1)}$) as the initial guess. 

The algorithm can be visualised as follows:

````python
def plot_secant(f, x_0, x_1, atol, ax, max_labels0=1, inset=True, ixmin=3.0, ixmax=3.2, 
                iymin=-0.1, iymax=0.1, zoom=8, loc0=1, loc1=3, loc2=2, maxiter=100, resfct=100, flabel=''):
    # variables to store to convergence pattern, initialise with two initial guess values
    x_n = [x_0, x_1] 
    y_n = [f(x_0), f(x_1)]    
    x_its = [x_0, x_1]
    # iterate secant method and store iteration convergence pattern
    i = 0
    # iterate until we hit break either as we hit tolerance or maximum number iterations
    while 1:     
        # Secant update
        x_zero = x_its[-1] - f(x_its[-1])*( (x_its[-1] - x_its[-2]) / (f(x_its[-1]) - f(x_its[-2])) )
        x_n.append(x_zero)        
        y_n.append(0.)
        x_its.append(x_zero)
        if abs( x_n[-1] - x_n[-2] ) < atol:
            break
        x_n.append(x_zero)
        y_n.append(f(x_zero))
        i = i+1
        if i >= maxiter:
            break

    # the iteration results
    ax.scatter(x_its, f(np.array(x_its)), marker='x', color='red', s=30)
    
    # the convergence pattern
    ax.plot(x_n, y_n, color='green', ls='--')
     
    # the function
    x = np.linspace( np.min(x_n), np.max(x_n), resfct)
    ax.plot(x, f(x), 'b')

    # zero line
    xlim = ax.get_xlim()
    ax.plot([xlim[0], xlim[1]], [0., 0.], 'k--')
    ax.set_xlim(xlim)

    # zoomed inset
    if inset:
        axins = zoomed_inset_axes(ax, zoom, loc=loc0)
        axins.scatter(x_its, f(np.array(x_its)), marker='x', color='red', s=30)
        axins.plot(x_n, y_n, color='green', ls='--')
        axins.plot(x, f(x), 'b')
        axins.plot([ixmin,ixmax],[0.,0.], 'k--')
        axins.set_xlim(ixmin, ixmax)
        axins.set_ylim(iymin, iymax)
        axins.get_xaxis().set_visible(False)
        axins.get_yaxis().set_visible(False)
        mark_inset(ax, axins, loc1=loc1, loc2=loc2, fc="none", ec="0.5")        
        
    ax.set_xlabel('$x$', fontsize=16)
    if not flabel:
        fl = '$f(x)$'
    else:
        fl = flabel
    ax.set_ylabel(fl, fontsize=16)
    ax.set_title('Secant method iteration', fontsize=20)


# case 1
def f(x):
    return x - np.exp(-x)


x0 = -1.
x1 = x0+0.75
fig, ax1 = plt.subplots(figsize=(10,6))
plot_secant(f, x0, x1, 1E-3, ax1, loc0=4, loc1=2, loc2=1,
                                            zoom=4, ixmin=0.4, ixmax=0.65, iymin=-0.2, iymax=0.2,
                                            flabel=r'$f(x) = x - \mathrm{e}^{-x}$')

# case 2
def f(x):
    return 2*x + x*np.sin(x-3) - 5


fig, ax1 = plt.subplots(figsize=(10,6))
x0 = 2.
x1 = x0+0.1
plot_secant(f, x0, x1, 1E-3, ax1, loc0=2, loc1=3, loc2=4, inset=True,
                                            zoom=4, ixmin=2.7, ixmax=2.85, iymin=-0.35, iymax=0.1,
                                            flabel=r'$f(x) = 2x + x\sin(x-3) - 5$')
````

# Potential convergence Issues [$\star$]

It's important to understand the ways some of the methods we've introduced can go wrong, and again to emphasise the value of a good starting guess.

Let's start by illustrating the basic concept of Newton root-finding methods for a well-behaved function

$$ f(x) := x^4 - 5. $$

````python
def f(x):
    return x**4 - 5

fig, ax1 = plt.subplots(figsize=(10,6))
x0 = 4.0
plot_newton(f, 1.e-8, x0, 1.e-7, ax1, loc0=2, loc1=3, loc2=4, zoom=5,
                                            ixmin=1.45, ixmax=1.85, iymin=-5, iymax=15, flabel=r'$f(x) = x^4 - 5$')
````

We've said several times now that the convergence of many algorithms often depends on the initial values provided to them. 

Complex equations, or systems thereof, provide plenty of situations that prevent convergence all together, lead to slow convergence, or cause convergence to an undesired root (e.g. one that is far from, and not the closest to, the initial guess). 

Their solution strongly depends on having a *good* initial guess. 

For example,

$$ f(x) = x\, \sin(\pi x) - e^{-x}, $$

provides for ample pitfalls, in particular for gradient-based methods (e.g. Newton and Secant).

We will illustrate these convergence issues below. 


They mostly fall into one of three categories:


1. Multiple roots in the vicinity - the algorithm converges to one of them, but this "one" could change with only slight changes to the initial guess.


2. Nearly singular/numerical overflow - the local gradient at guess $x^{(k)}$ is near zero, so that $x^{(k+1)}$, the intersection of the local gradient approximation with the $x$ axis is beyond the representable range. Even if we don't get overflow, we can still jump to a location a long distance away from our initial guess and end up converging to the wrong root.


3. Oscillations - the algorithm bounces back and forth between (near) identical $x^{(k)}$.

## Multiple roots

For many functions multiple roots exist, and the algorithm depends on and can be sensitive to the initial guess.

Let's see an example where we choose two initial guesses (0.0 and 0.1) and visualise how this relatively small difference leads to convergence to different roots.

````python
def f(x):
    return x*np.sin(np.pi*x)-np.exp(-x)

fig, ax1 = plt.subplots(figsize=(10,6))
x0 = 0.0
plot_newton(f, 1.e-8, x0, 1.e-7, ax1, inset=False, flabel='$f(x)= x \mathrm{sin}(\pi x) - \mathrm{e}^{-x}$')
````

# Nonlinear systems of equations [$\star\star$]

Now we assume we have a problem in the form

$$\boldsymbol{f}\,(\boldsymbol{x}) = \boldsymbol{0}.$$

The is the *vector form* of the problem which can be written out in scalar form as

$$\begin{align*}
f_0 (x_0,x_1,\ldots x_n) & = 0, \\[5pt]
f_1 (x_0,x_1,\ldots x_n) & = 0, \\[5pt]
& \vdots \\[5pt]
f_n (x_0,x_1,\ldots x_n) & = 0,
\end{align*}$$

i.e. we have $n+1$ simultaneous equations for $n+1$ unknowns.

If all of these equations were linear then we would be back in the situation we were in in lecture 3 (and could rewrite the above system as $\boldsymbol{0} = \boldsymbol{f}\,(\boldsymbol{x}) \equiv A \boldsymbol{x} - \boldsymbol{b}$).

If at least one of these equations are nonlinear then we can't formulate this as a simple matrix problem to invert.

Newton's method represents a popular method to tackle this problem. 

Let's extend our (scalar) derivation from earlier:

The [multi-dimensional version of Taylor series](https://en.wikipedia.org/wiki/Taylor_series#Taylor_series_in_several_variables) take the form (here assuming a scalar function, $f_i$, of a vector input $\boldsymbol{x}:=(x_0,\ldots,x_n)$)

$$ f_i(\boldsymbol{x} + \Delta\boldsymbol{x}) = f_i(\boldsymbol{x}) + \sum_{j=0}^n \frac{\partial f_i}{\partial x_j} \Delta x_j+ \mathcal{O}(\Delta\boldsymbol{x})^2. $$

Dropping the second-order terms, and combining the above expression for all the functions in our system into vector notation, yields

$$ \boldsymbol{f}\,(\boldsymbol{x} + \Delta\boldsymbol{x}) = \boldsymbol{f}\,(\boldsymbol{x}) + J(\boldsymbol{x})\Delta\boldsymbol{x},$$

where $J(\boldsymbol{x})$ is called the [*Jacobian matrix*](https://en.wikipedia.org/wiki/Jacobian_matrix_and_determinant) - an $(n+1)\times (n+1)$ matrix made up of the partial derivatives:

$$ J_{ij} := \frac{\partial f_i}{\partial x_j}. $$

Now suppose we have a solution $\boldsymbol{x}^{(k)}$ in our iteration and we want an updated (hopefully improved) value $\boldsymbol{x}^{(k+1)}$. 

The Taylor series analysis tells us that:

$$ \boldsymbol{f}\,(\boldsymbol{x}^{(k+1)}) = \boldsymbol{f}\,(\boldsymbol{x}^{(k)}) + J(\boldsymbol{x}^{(k)})(\boldsymbol{x}^{(k+1)} - \boldsymbol{x}^{(k)}).$$

As before, we want $\boldsymbol{f}\,(\boldsymbol{x}^{(k+1)}) \approx \boldsymbol{0}$ so specify this in the above to yield

$$ J(\boldsymbol{x}^{(k)})(\boldsymbol{x}^{(k+1)} - \boldsymbol{x}^{(k)}) = -\boldsymbol{f}\,(\boldsymbol{x}^{(k)}),$$

or with the notation $\Delta\boldsymbol{x}^{(k)} :=  \boldsymbol{x}^{(k+1)} - \boldsymbol{x}^{(k)}$ if we can solve 

$$ J(\boldsymbol{x}_k)\Delta\boldsymbol{x}^{(k)} = -\boldsymbol{f}\,(\boldsymbol{x}^{(k)}),$$

[Note this is a *linear* system since $J(\boldsymbol{x}_k)$ is simply a matrix with known entries at each iteration for a fixed and known $\boldsymbol{x}^{(k)}$]

for the update $\Delta\boldsymbol{x}^{(k)}$, then we immediately have the iteration update

$$\boldsymbol{x}^{(k+1)} := \boldsymbol{x}^{(k)} + \Delta\boldsymbol{x}^{(k)}. $$

We could of course also have just written

$$\boldsymbol{x}^{(k+1)} := \boldsymbol{x}^{(k)} - [J(\boldsymbol{x}^{(k)})]^{-1}\boldsymbol{f}\,(\boldsymbol{x}^{(k)}),  $$

but didn't do so to emphasise the point that we don't really want to actually form the inverse matrix in our implementation. Rather we want to call a linear solver to solve for $\Delta\boldsymbol{x}^{(k)}$.

So the iteration is the same as it was for the scalar case. However, the implementation is complicated by two factors:


- (1) We need to solve a linear system at each iteration of the nonlinear solver, but of course we know how to do this from the last lecture! (Note that if this linear solver was also itself iterative in nature we talk about *inner* and *outer* iterations to distinguish the iterations of the linear solver from those of the nonlinear loop which sits above it).


- (2) To actually solve this linear system, we need the matrix $J(\boldsymbol{x}^{(k)})$. Just as in our Newton codes for scalar problems above where we had two versions, one where we passed in a function describing the derivative, and another where we numerically approximated it, we can also perform a numerical approximation to obtain the Jacobian.
