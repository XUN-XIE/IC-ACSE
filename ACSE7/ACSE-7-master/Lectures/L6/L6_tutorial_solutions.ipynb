# Optimal Box Shape
A chocolate manufacturer wants to design a 6-sided box: six sides and a regular hexagon as the top and bottom:

<img width=200 src="https://images-na.ssl-images-amazon.com/images/I/41LOrfo46sL._AC_SL1000_.jpg">

For a given volume, what is the optimal ratio between the height $h$, and the radius $r$ of the hexagon in order to minimze the amount of packaging, i.e. to minimize the area of the box. **Hint:** you were given the formula of the area of a regular hexagon in the tutorial for lecture 4.

### Solution <a class="tocSkip"></a>

We first derive the following formulas for the area and volume of the box. The area of the hexagon is $\tfrac 12 r^2 6 sin(\pi/3)=\frac 32\sqrt{3}r^2$): The hexagon can divided in 6 triangles, which because the internal angle is 360/6=60, are equilateral, and thus the edges of the hexagon have length $r$. This gives:

$$
  A(h,r) = 6 h r + 3\sqrt{3}r^2 \\
  V(h,r) = \frac 32\sqrt{3}r^2 h
$$

We write as a standard constrained optimisation problem:

$$
  \text{minimize } f(h,r) = A(h,r) = 6 h r + 3\sqrt{3}r^2, \\
  \text{subject to } g(h,r) = V(h,r) - V_0 = \frac 32\sqrt{3}r^2 h - V_0 = 0
$$

We define a Lagrangian $L(h,r, \lambda) = f(h,r) + \lambda g(h,r)$ and the associated
equations to find a stationary point of the Lagrangian:

$$
  \frac{\partial L}{\partial h} = 6 r  + \lambda \frac 32\sqrt{3}r^2 = 0, \\
  \frac{\partial L}{\partial r} = 6 h + 6\sqrt{3}r + \lambda 3\sqrt{3}r h = 0, \\
  \frac{\partial L}{\partial\lambda} = \frac 32\sqrt{3}r^2 h - V_0 = 0.
$$

From the first equation we may derive

$$
  \lambda = -\frac{4}{\sqrt{3}r}.
$$

If we substitute this into the second equation, we get

$$
  6 h + 6\sqrt{3}r -12 h = 0 \\
  h = \sqrt{3}r
$$

in other words the height $h$ should $\sqrt{3}$ times the radius $r$.

# Constrained Problem

Consider the following constrained problem:

$$
  \text{minimize}f(x, y) = x^4 + y^4 + 2x^2 \\
  \text{subject to }x^2 + y ^2 = 2
$$

* sketch the feasible region
* is this a convex constrained optimisation problem?
* find the stationary points of the constrained problem

### Solution <a class="tocSkip"></a>

The feasible region is the circle with radius $\sqrt{2}$ around the origin. This is *not* a convex optimisation problem, both $f$ and $g$ (defined) below are convex functions, but we have defined a convex optimisation problem as having linear constraints which these are not. Indeed the circle is not a convex set. We may therefore not draw any conclusions about the stationary points we find for the constrained problem.

We define a lagrangian

$$
  L(x,y) = f(x,y) + \lambda g(x, y)
$$

and derive the usual equations for stationary points of the Lagrangian:

\begin{align}
  \frac{\partial}{\partial x} f(x,y) + \lambda \frac{\partial}{\partial x} g(x,y) = 0 &&
    4x^3 + 4x + 2\lambda x =0 & \quad(1) \\
  \frac{\partial}{\partial y} f(x,y) + \lambda \frac{\partial}{\partial y} g(x,y) = 0 &&
    4y^3 + 2\lambda y = 0  & \quad(2) \\
  \frac{\partial}{\partial\lambda} L(x,y,\lambda) = g(x,y) = 0  &&   
x^2 + y^2 = 2  & \quad(3)
\end{align}

If $x=0$ to satisfy eqn. 1, then $y=\pm\sqrt{2}$ (3rd eqn.), and thus $\lambda=-4$ (2nd eqn.)

If $y=0$ to satsify eqn. 2, then $x=\pm\sqrt{2}$ (3rd eqn.), and thus $\lambda=-6$ (1st eqn.)

If $x\neq 0$ and $y\neq 0$, we can simplify:

$$
  4 x^2 + 4 + 2\lambda = 0 \\
  4 y^2 + 2\lambda = 0 \\
  x^2 + y^2 = 2
$$

Adding the 1st and 2nd eqn., we get:

$$
  4 x^2 + 4 y^2 + 4 + 4\lambda = 8 + 4 + 4\lambda \implies \lambda = -3
$$

and we can derive the folllowing four solutions:

$$
 (x,y) = (-\sqrt{\tfrac 12}, -\sqrt{\tfrac 32}),\quad
 (x,y) = (\sqrt{\tfrac 12}, -\sqrt{\tfrac 32}),\quad
 (x,y) = (-\sqrt{\tfrac 12}, \sqrt{\tfrac 32}),\quad
 (x,y) = (\sqrt{\tfrac 12}, \sqrt{\tfrac 32}),\quad
$$

**In summary:** we have found 8 stationary points:

\begin{align*}
 (x,y) = (0,-\sqrt{2}) \quad
 (x,y) = (0,\sqrt{2}) \quad
 (x,y) = (-\sqrt{2}, 0) \quad
 (x,y) = (\sqrt{2}, 0) \\
 (x,y) = (-\sqrt{\tfrac 12}, -\sqrt{\tfrac 32}),\quad
 (x,y) = (\sqrt{\tfrac 12}, -\sqrt{\tfrac 32}),\quad
 (x,y) = (-\sqrt{\tfrac 12}, \sqrt{\tfrac 32}),\quad
 (x,y) = (\sqrt{\tfrac 12}, \sqrt{\tfrac 32}),\quad
\end{align*}

The result might seem surprising - the function $f(x,y)=x^4+y^4+2x^2$ on its own is a smooth convex function, with only a single local minimum (as do all convex functions). In the following figure we have plotted the intersection of the graph of $f$ with the circle $x^2+y^2=2$. For clarity we have only plotted half of the figure, the half for which $x+y>0$:

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1, 1, 1, projection='3d')
r = np.sqrt(2)*1.1
ax.axis([-r, r, -r, r])
xr = np.linspace(-r, r, 50)
yr = np.linspace(-r, r, 50)
# create a Nx x Ny x 2 (Nx=Ny=50) array of grid points
xy = np.array(np.meshgrid(xr, yr)).T
x = xy[:,:,0]
y = xy[:,:,1]
f = x**4 + y**4 + 2*x**2

# this is a hack to ensure only the x+y>0 halve is plotted
# it causes a warning, but seems to work
f[x+y<0] = np.nan

ax.plot_surface(xy[:,:,0], xy[:,:,1], f, alpha=0.8)
ax.set_xlabel('x-axis', size=24, labelpad=10)
ax.set_ylabel('y-axis', size=24, labelpad=10);
ax.view_init(elev=5, azim=-100)

# now paramaterize the x^2+y^2=2 semi-circle
# using (x,y)=sqrt(2)*[cos(phi), sin(phi)] for -pi/4<phi<3*pi/4
pi = np.pi
x = []; y = []
phir = np.linspace(-pi/4,3*pi/4, 100)
for phi in phir:
    x.append(np.sqrt(2)*np.cos(phi))
    y.append(np.sqrt(2)*np.sin(phi))
x = np.array(x)
y = np.array(y)

ax.plot(x, y, 0, '-', color='grey')  # semi-circle in the z=0 plane
ax.plot(x, y, x**4+y**4+2*x**2, '-', color='black')  # the function values (x,y,f(x,y))

# the stationary points we have found ourselves, for which x+y>0:
for xi, yi in [[-np.sqrt(.5), np.sqrt(1.5)], [np.sqrt(.5), np.sqrt(1.5)], [0, np.sqrt(2)], [np.sqrt(2), 0]]:
    print(xi, yi)
    ax.plot([xi], [yi], [xi**4+yi**4+2*xi**2], 'o', color='red')
```

# Derivation of the Trust Region method
Using the techniques from today's lecture, we can derive the formula that modifies the Hessian in the Trust Region method (lecture 3).

1. Given a general quadratic function of the usual form

$$
  f(\vec x) = \frac 12 \vec x^T \mat A\vec x - \vec b^T\vec x +c
$$

find an equation for the minimum $f(\vec x)$ for points $\vec x$ *on* the circle (in 2D) or sphere (in 3D) or "hyper-sphere" in n dimensions using the following equality constraint:

$$
  g(\vec x) = \frac 12\vec x^T \vec x - \frac 12 r^2 = 0
$$

You don't need to solve for the Lagrange multiplier (for this see the "final note" below), but try arrive an equation in the following form:

$$
  \left[\mat A + \lambda\mat I\right]\vec x = 
\vec b
$$

where $\mat I$ is the identity matrix. You may assume that $\mat A$ is SPD to ensure that a minimum exists, and is unique.

### Solution <a class="tocSkip"></a>

We write a Lagrangian:

$$
  L(\vec x, \lambda) = f(\vec x) + \lambda g(\vec x)
$$

and derive the relevant equations by looking for a stationary point of $L$:

\begin{align*}
  \frac{\partial}{\partial\vec x} L(\vec x, \lambda)
  &= \frac{\partial}{\partial\vec x} f(\vec x) + \lambda \frac{\partial}{\partial\vec x}g(\vec x) \\
  &= \mat A\vec x - \vec b + \lambda \vec x =0 \\
  &= \mat A\vec x - \vec b + \lambda \mat I\vec x =0 
  \implies \left[\mat A + \lambda\mat I\right]\vec x = 
\vec b
  \frac{\partial}{\partial\lambda} L(\vec x, \lambda) &= g(\vec x) = 0
\end{align*}

The first is the asked for result. The second equation is again the equality constraint.

2. Now proof the following expression for the optimisation problem of finding the minimum *inside* (or on the boundary of) a circle/sphere/hyper-sphere of radius $r$:

$$
  \left[\mat A + \mu\mat I\right]\vec x = 
\vec b
$$

What conditions do we have for $\mu$?

### Solution <a class="tocSkip"></a>

The derivation is nearly the same as before. We now have an inequality constraint:

$$
  h(\vec x)=\tfrac 12\vec x^T\vec x - \tfrac 12r^2 \leq 0
$$

So we write down a Lagrangian:

$$
  L(\vec x, \mu) = f(\vec x) + \mu h(\vec x)
$$

and we need to satisfy the following KKT equations:

\begin{align*}
  \frac{\partial L(\vec x, \mu)}{\partial\vec x} &= 0 \\  
  h(\vec x) & \leq 0 \\
  \mu &\geq 0 \\
  \mu h(\vec x) &= 0 
\end{align*}

By the last equation, there are two cases to consider: $\mu=0$, minimum inside the sphere; then the first equation becomes:

\begin{align*}
  \frac{\partial L(\vec x, \mu)}{\partial\vec x} = 0 \\  
  \frac{\partial f(\vec x)}{\partial\vec x} = 0 \\  
\end{align*}

in other words, we simply searching for a stationary point of $f(\vec x)$, which implies

$$
  \mat A\vec x = b
$$

which is consistent with the asked for expression (when $\mu=0$).

When $\mu>0$, this implies by the last equation that $h(\vec x)=0$, in other words, we find a stationary point on the sphere (the boundary). That case is exactly the same as the previous question with $\lambda$ replaced by $\mu$ which results in the same equation as before.

Thus, both for $\mu=0$ and $\mu>0$ we obtain the desired result, and the conditions are $\mu\geq 0$ and $\mu h(\vec x)=0$.

3. Write down the first three terms in the Taylor series approximation of a function $f:\R^n \to\R$ in a point $\vec x^{(i)}+\vec p$ near another point $\vec x^{(i)}$:

$$
  f(\vec x^{(i)}+\vec p) = ...
$$

Find the minimum of this approximation formed by the three terms (derive an expression to solve for $\vec p$), and relate your result to the (unconstrained) Newton method.

### Solution <a class="tocSkip"></a>
The Taylor series approximation reads:

$$
  f(\vec x^{(i)}+\vec p) = f(\vec x^{(i)}) +
  f'(\vec x^{(i)})^T \vec p + \frac 12\vec p^T f''(\vec x^{(i)})\vec p
  + \mathcal(O)(\|\vec p\|^3)
$$

The first three terms for a quadratic approximation which can be related to the standard form

$$
  f_{\text{quad}}(\vec p) = \frac 12\vec p^T\mat A\vec p
  - \vec b^T\vec p + c
$$

where $\mat A=f''(\vec x^{(i)}), \vec b=-f'(\vec x^{(i)}),$ and
$c=  f(\vec x^{(i)})$. The solution for the minimum of this quadratic approximation is therefore:

$$
  \mat A\vec x = \vec b \implies f''(\vec x^{(i)})\vec p = -f'(\vec x^{(i)})
$$

which is exactly the linear system that gets solved in the Newton method to compute the step vector $\vec p$.

4. In the trust region method, we first try a standard Newton step, which you've derived again in the previous question. If the resulting $\vec p$ does not lead to a sufficient decrease in the new function value $f(\vec x^{(i)}+\vec p)$, we restrict the region in which we *trust* the quadratic approximation. We do this by choosing a radius $r$ that is smaller than the step size of our first try, and find the minimum of the quadratic approximation with the restriction that $\|\vec p\|\leq r$. In other words we choose a sphere of radius $r$ around $\vec x^{(i)}$ in which we trust the approximation. Combine your answers to question 2. and 3. to derive an equation for this trust region approach. Again you may assume that the Hessian in $\vec x^{(i)}$ is SPD. Why can we assume that the minimum is *not* going to be found in the interior of the radius-$r$ sphere?

### Solution <a class="tocSkip"></a>
In the answer to question 3 we saw that the quadratic approximation can be related to the standard quadratic form with $\mat A=f''(\vec x^{(i)}), \vec b=-f'(\vec x^{(i)})$. Substituting this in our previous result
we get

$$
  \left[\mat A + \mu\mat I\right]\vec p = 
  \left[f''(\vec x^{(i)}) + \mu\mat I\right]\vec p = 
-f'(\vec x^{(i)})
$$

The reason we may assume that the minimum is not in the interior, but on the boundary of the sphere is that we know that $\mat A$ is SPD. This means that $f$ only has a single global minimum. Our first try (with a standard Newton step), would have brought us to this minimum. Because we have chosen $r$ to be smaller than that step, we know the minimum is not within the sphere.

Because of this we may assume $\mu>0$ (i.e strictly bigger than zero), and $h(\vec p)=0$, or $\|\vec p\|=r$.

### Final note <a class="tocSkip"></a>
In our previous answer, like in the answer to 2. we still have a $\mu$ that we don't know. We can derive an equation for $\mu$ by rewriting the result of answer 2 in the form:

$$
  \vec x = \left[\mat A + \mu\mat I\right]^{-1}\vec b
$$

If we assume the minimum is found on the boundary of the sphere, we have

$$
  \vec x^T\vec x = r^2
$$

which we can work out to

\begin{align*}
  \vec b^T \left[\mat A + \mu\mat I\right]^{-1,T}\left[\mat A + \mu\mat I\right]^{-1}\vec b &= r^2 \\
  \vec b^T \left(\left[\mat A + \mu\mat I\right]\left[\mat A + \mu\mat I\right]^T\right)^{-1}\vec b &= r^2
\end{align*}

This is an equation that we can solve for $\mu$. As you can see, it is not a very nice equation to solve. There are therefore various approaches to deal with this situation:

* in an *exact trust region approach* we do solve the equation above. We choose an initial radius $r$, which we may for instance halve every time we find a point that does not satisfy a sufficient decrease condition (e.g. Armijo). We then solve the above equation for $\mu$. It is a one-dimensional nonlinear equation, so in principle we can just solve with for instance Newton. Obviously this might add considerable expense to the algorithm
* in other approaches, instead of choosing a $r$ beforehand, we simply treat $\mu$ as a free parameter. We start using $\mu=0$ which gives plain Newton, then if the sufficient decrease hasn't been achieved we simply increase $\mu$. For very big $\mu$ we get

$$
  \mat A+\mu\mat I\approx \mu\mat I
$$

and therefore we have a update step of

$$
  \mu\mat I\vec p = -f'(\vec x^{(I)}) \implies \vec p =-\frac 1\mu f'(\vec x^{(I)})
$$

which is just Steepest Descent with a "line search parameter" $\lambda=1/\mu$. Thus we can choose any $\nu$ between zero and infinity, where for big $\mu$ we transform the method into Steepest Descent. So we can use this method in a similar way as backtracking line search, but instead of using a $\lambda$ that starts at 1 and is decreased until we satisfy a sufficient decrease condition in $f$, we now have a parameter $\nu$ that we start at zero and keep increasing until the condition is satisfied. For more information see e.g. [Kelley's "Iterative Methods for Optimization"](https://www.amazon.com/Iterative-Methods-Optimization-Frontiers-Mathematics/dp/0898714338).
